# SFT Dataset Generation — from `passed_mcqs.jsonl` (S5 Output)

**AI Tutor System — Teacher-Student Knowledge Distillation**

---

## Purpose

This notebook builds the final **Supervised Fine-Tuning (SFT) dataset** for Qwen student-model training.  
It reads from the **S5 validation pipeline output** (`passed_mcqs.jsonl`) — the
highest-quality, fully-validated MCQs that survived all five validation stages — and
converts each slot-level record into a **LLaMA-Factory–compatible SFT training sample**.

---

## Input → Output

| | File | Format | Contents |
|---|---|---|---|
| **Input** | `passed_mcqs.jsonl` | JSONL | S5 stage-by-stage `passed.jsonl` records (one per validated MCQ slot) |
| **Input** | `selected_subset_v3.jsonl` | JSONL | S3 chunks (provides `metadata`, `keywords`, `context_fringe`) |
| **Output** | `sft_metadata_conditioned.jsonl` | JSONL | Metadata-conditioned SFT records (Option A) |
| **Output** | `sft_dataset_info.json` | JSON | LLaMA-Factory `dataset_info.json` entries |

---

## Input Record Schema (`passed_mcqs.jsonl`)

Each line in `passed_mcqs.jsonl` is a S5 stage-by-stage `passed.jsonl` record:

```json
{
  "mcq_data": {
    "chunk_id":   "<str>",
    "text":       "<str>  — the source chunk text",
    "slot_index": "<int>",
    "chunk_type": "<str>  — definition | process | comparison | example | argument",
    "concept":    "<str>  — the concept this slot tests",
    "bloom_level":"<str>  — remember | understand | apply | analyze | evaluate | create",
    "mcq": {
      "question":    "<str>",
      "options":     {"A": "<str>", "B": "<str>", "C": "<str>", "D": "<str>"},
      "answer":      "<A|B|C|D>",
      "explanation": "<str>"
    }
  },
  "chunk_id":    "<str>",
  "slot_index":  "<int>",
  "stage_name":  "llm_judge",
  "stage_number": 5,
  "timestamp":   "<ISO-8601 str>",
  "validation_result": {
    "passed": true,
    "reason": "",
    "scores": { "correctness": 4.0, "clarity": 4.0, "distractors": 3.5,
                "educational_value": 4.0, "chunk_grounding": 4.0, "overall": 3.9 },
    "ambiguous": false
  },
  "scores": {}
}
```

---

## SFT Format: `sharegpt` (LLaMA-Factory)

Every training record uses the **`sharegpt`** conversation format with three turns:

```
system   → static domain expert persona + output format rules
user     → chunk text + metadata conditioning tags + task instruction
assistant → the validated MCQ as a clean JSON object
```

This matches the **runtime inference format exactly** — no train/inference mismatch.

---

## Two Training Variants

| Variant | User turn includes metadata? | Purpose |
|---|---|---|
| **Option A** `sft_metadata_conditioned.jsonl` | Yes — `[BLOOM LEVEL]`, `[CHUNK TYPE]`, `[FOCUS CONCEPT]` | Primary training target; enables runtime Bloom control |

## 3. Load Inputs

### 3.1  Load S5 Passed MCQs

Each record is a S5 stage-by-stage `passed.jsonl` entry.  
The validated MCQ and its conditioning metadata live inside `mcq_data`.

````python
def load_jsonl(path: str) -> List[Dict]:
    """Load a JSONL file, skipping blank lines and malformed records."""
    records: List[Dict] = []
    skipped = 0
    with open(path, encoding="utf-8") as f:
        for lineno, line in enumerate(f, start=1):
            line = line.strip()
            if not line:
                continue
            try:
                records.append(json.loads(line))
            except json.JSONDecodeError as exc:
                print(f"  [WARN] Skipping malformed line {lineno}: {exc}")
                skipped += 1
    if skipped:
        print(f"  Skipped {skipped} malformed line(s) in {os.path.basename(path)}")
    return records


s5_records: List[Dict] = load_jsonl(S5_PASSED_PATH)
print(f"Loaded {len(s5_records):,} passed MCQ records from S5.")
````

### 3.2  Load S3 Chunks

S3 chunks provide the **keywords** and **context_fringe** fields that were stored
in the original chunk but are not carried through the S5 slot-level records.

If a chunk_id from S5 is not found in S3, the SFT record is still built using
only the data available in `mcq_data` (graceful degradation — keywords default
to empty, context fringe defaults to absent).

````python
# Build lookup: chunk_id → full S3 chunk dict
s3_lookup: Dict[str, Dict] = {}
for rec in load_jsonl(S3_CHUNKS_PATH):
    cid = rec.get("chunk_id")
    if cid:
        s3_lookup[cid] = rec

print(f"S3 chunk lookup : {len(s3_lookup):,} chunks")

# Coverage check
s5_chunk_ids  = {r.get("chunk_id", r.get("mcq_data", {}).get("chunk_id", ""))
                 for r in s5_records}
matched       = s5_chunk_ids & set(s3_lookup.keys())
unmatched     = s5_chunk_ids - set(s3_lookup.keys())

print(f"Unique chunk_ids in S5 : {len(s5_chunk_ids):,}")
print(f"Matched in S3          : {len(matched):,}")
if unmatched:
    print(f"NOT matched (keywords/fringe will be empty): {len(unmatched):,}")
    for cid in list(unmatched)[:5]:
        print(f"  {cid}")
````

## 4. Record Extraction Helper

`extract_slot_data()` normalises a single S5 passed record into a clean,
flat dict containing only the fields needed for SFT construction.

**Fields kept:**

| Field | Source | Purpose |
|---|---|---|
| `chunk_id` | `mcq_data` | Record identity / dedup key |
| `slot_index` | `mcq_data` | Slot identity within the chunk |
| `text` | `mcq_data` | Source chunk text → user turn |
| `bloom_level` | `mcq_data` | Bloom conditioning tag |
| `chunk_type` | `mcq_data` | Chunk-type conditioning tag |
| `concept` | `mcq_data` | Focus concept conditioning tag |
| `keywords` | S3 lookup | Distractor-quality hint in user turn |
| `prev_sentence` | S3 lookup | Context fringe (background only) |
| `next_sentence` | S3 lookup | Context fringe (background only) |
| `mcq` | `mcq_data` | The validated MCQ → assistant turn |

**Fields excluded:**

| Field | Reason for exclusion |
|---|---|
| `file_id` | Internal pipeline routing; irrelevant to model training |
| `s5_validation` | Validation bookkeeping; quality is implicit in the curated dataset itself |
| `chunk_type_confidence` | Classifier confidence; not meaningful in a training context |
| `stage_name`, `stage_number` | Pipeline provenance; not a training signal |
| `timestamp` | Pipeline provenance; not a training signal |
| `scores` (top-level) | Redundant empty dict from S5 per-MCQ records |
| `validation_result` | Pipeline bookkeeping; removed entirely from SFT output |

````python
def extract_slot_data(
    s5_record: Dict,
    s3_lookup: Dict[str, Dict],
) -> Optional[Dict]:
    """
    Extract and normalise training-relevant fields from one S5 passed record.

    Parameters
    ----------
    s5_record : one line from passed_mcqs.jsonl
    s3_lookup : chunk_id -> S3 chunk dict (for keywords + context_fringe)

    Returns
    -------
    Flat dict with exactly the fields needed for SFT construction,
    or None if the record is structurally invalid.
    """
    # ── 1. Pull mcq_data sub-dict ─────────────────────────────────────────────
    mcq_data: Dict = s5_record.get("mcq_data", {})
    if not mcq_data:
        return None

    # ── 2. Required fields from mcq_data ─────────────────────────────────────
    chunk_id  : str       = mcq_data.get("chunk_id", "").strip()
    slot_index: int       = mcq_data.get("slot_index", 0)
    text      : str       = mcq_data.get("text", "").strip()
    bloom     : str       = mcq_data.get("bloom_level", "understand").strip().lower()
    chunk_type: str       = mcq_data.get("chunk_type", "definition").strip().lower()
    concept   : str       = mcq_data.get("concept", "").strip()
    mcq       : Dict      = mcq_data.get("mcq", {})

    # Validate the minimum required fields
    if not chunk_id or not text or not mcq:
        return None

    # Validate MCQ sub-structure
    if not mcq.get("question") or not mcq.get("options") or not mcq.get("answer"):
        return None

    if set(mcq.get("options", {}).keys()) != {"A", "B", "C", "D"}:
        return None

    # ── 3. Enrich from S3 lookup (keywords + context fringe) ─────────────────
    s3_chunk    : Dict        = s3_lookup.get(chunk_id, {})
    s3_meta     : Dict        = s3_chunk.get("metadata", {})
    fringe      : Dict        = s3_chunk.get("context_fringe", {})

    keywords    : List[str]   = s3_meta.get("keywords", [])
    prev_sentence: str        = (fringe.get("prev_sentence") or "").strip()
    next_sentence: str        = (fringe.get("next_sentence") or "").strip()

    # ── 4. Build clean MCQ dict (only the 4 training-relevant fields) ─────────
    clean_mcq = {
        "question":    mcq.get("question", "").strip(),
        "options":     mcq.get("options", {}),
        "answer":      mcq.get("answer", "").strip().upper(),
        "explanation": mcq.get("explanation", "").strip(),
    }

    # ── 5. Return flat, clean dict ────────────────────────────────────────────
    return {
        "chunk_id":     chunk_id,
        "slot_index":   slot_index,
        "text":         text,
        "bloom_level":  bloom,
        "chunk_type":   chunk_type,
        "concept":      concept,
        "keywords":     keywords,
        "prev_sentence": prev_sentence,
        "next_sentence": next_sentence,
        "mcq":          clean_mcq,
    }


print("extract_slot_data() defined.")
````

## 5. Prompts — Preserved from S4 Generation Pipeline

The system prompt and user prompt structure are **preserved exactly** from the
original S4 generation pipeline so the student model's training format matches
its inference format precisely.

### Changes from the original `build_sft_record()`

| What changed | Why |
|---|---|
| Input source is `passed_mcqs.jsonl` (S5), not `s4_success.jsonl` | Uses only the highest-quality, fully-validated MCQs |
| One SFT record per **slot** (not per chunk) | Training signal is per MCQ, not per chunk |
| `sharegpt` conversation format with `system`/`user`/`assistant` turns | Direct LLaMA-Factory compatibility; matches runtime inference format |
| Assistant turn is a **clean JSON object** (not wrapped in a JSON array) | One slot = one MCQ = one JSON object; no array needed |
| Metadata tags use natural-language bracket format `[BLOOM LEVEL]` | Better adherence in instruction-tuned models than raw JSON metadata |
| `
| `file_id`, `s5_validation`, `chunk_type_confidence` fully removed | Pipeline metadata; zero training value |


````python
# ── System Prompt ─────────────────────────────────────────────────────────────
# Preserved from the S4 generation pipeline (SYSTEM_PROMPT).
# Kept as a static string — identical across all training records.
# NOT repeated in the user turn (matches how S4 sent it as system_instruction).

SFT_SYSTEM_PROMPT = "\n".join([
    # -- Persona & core behavior
    "You are an expert educational content designer for Bloom's Taxonomy-aligned assessment.",
    "Generate multiple-choice questions from educational text.",
    "Return ONLY valid JSON. No markdown fences, preambles, or commentary.",
    "",

    # -- Bloom levels (compressed)
    "## BLOOM LEVELS",
    "remember: recall facts | understand: explain meaning | apply: use in scenario",
    "analyze: compare/reason | evaluate: justify decision | create: design/propose",
    "",

    # -- Chunk type forms (compressed)
    "## CHUNK TYPE FORMS",
    "definition: explain concept | process: steps/sequence | comparison: differences",
    "example: real case | argument: evaluate claim",
    "",

    # -- Grounding rule (highest priority)
    "## GROUNDING (highest priority)",
    "- Correct answer MUST be explicitly stated in or directly inferable from the chunk.",
    "- Do NOT use external knowledge.",
    "",

    # -- MCQ quality rules
    "## MCQ RULES",
    "- Exactly 4 options: A, B, C, D. Exactly ONE correct answer.",
    "- Forbidden: 'All of the above', 'None of the above', 'Both A and B'.",
    "- Options must be similar in length (±25-30%) and share the same grammatical structure.",
    "- Options must be mutually exclusive. Do NOT embed the correct answer text in the question.",
    "",

    # -- Distractor rules
    "## DISTRACTORS",
    "- Use the keywords list to craft plausible distractors.",
    "- All 3 wrong options must belong to the same category/domain as the correct answer.",
    "- Distractors must be plausible to a student who skipped the chunk.",
    "",

    # -- Question phrasing
    "## QUESTION PHRASING",
    '- Do NOT say "according to the text", "based on the chunk", or "as stated above".',
    "- Question stem must be self-contained and <= 50 words.",
    "",

    "- You MUST complete the JSON fully. Do not stop mid-output.",
    "- Ensure all 4 options A,B,C,D are present before finishing.",
    "",

    "## EXPLANATION RULES",
    "- Do NOT mention option letters (A, B, C, D).",
    "- Refer to the answer by its content, not its position.",
    "- Do NOT use phrases like 'according to the text' or 'according to the chunk'.",
    "- Justify the correct concept clearly using its meaning.",
    "",

    # -- Output format (single MCQ object)
    "## OUTPUT FORMAT — return ONLY this JSON object:",
    "{",
    '  "question": "<full question stem>",',
    '  "options": {"A": "...", "B": "...", "C": "...", "D": "..."},',
    '  "answer": "<A|B|C|D>",',
    '  "explanation": "<1-2 sentences: why correct, referencing chunk>"',
    "}",
])

print(f"SFT_SYSTEM_PROMPT  : {len(SFT_SYSTEM_PROMPT):,} chars  (~{len(SFT_SYSTEM_PROMPT)//4} tokens)")
````

````python
# ── Bloom and Chunk-Type instruction dicts ───────────────────────────────────
# Preserved from the S4 pipeline (BLOOM_INSTRUCTION, CHUNK_TYPE_FORM).
# Used to build the

BLOOM_INSTRUCTION: Dict[str, str] = {
    "remember"  : "Generate a RECALL question. Test a fact, definition, or named item directly stated in the text.",
    "understand": "Generate an INTERPRETATION question. Test whether the student can explain or paraphrase.",
    "apply"     : "Generate a SCENARIO question. Place the concept in a new, concrete situation.",
    "analyze"   : "Generate a COMPARISON or REASONING question. The stem MUST compare two items or ask for a cause.",
    "evaluate"  : "Generate a JUDGMENT question. Ask the student to choose and justify.",
    "create"    : "Generate a DESIGN or SYNTHESIS question.",
}

CHUNK_TYPE_FORM: Dict[str, str] = {
    "definition": "Use 'What is / Which best describes' style.",
    "process"   : "Use 'What happens when / Which step follows' style.",
    "comparison": "Use 'How does X differ from Y / Which distinguishes' style.",
    "example"   : "Use a scenario-based style embedding the concept in a concrete case.",
    "argument"  : "Require evaluating a claim, premise, or conclusion.",
}

print("BLOOM_INSTRUCTION and CHUNK_TYPE_FORM defined.")
````

````python
# ── User Prompt Builders ──────────────────────────────────────────────────────

def build_user_prompt_conditioned(slot: Dict) -> str:
    """
    Build the user turn for OPTION A (metadata-conditioned) SFT records.

    Structure (preserved from S4 build_user_prompt, extended with structured tags):

      ## CHUNK TEXT          — the source text
      ## CONTEXT             — context fringe (prev/next sentences, background only)
      ## METADATA            — chunk_type + keywords  (from S4)
      ---
      [BLOOM LEVEL]          — explicit conditioning tag
      [CHUNK TYPE]           — explicit conditioning tag
      [FOCUS CONCEPT]        — explicit conditioning tag
      ---
      ## YOUR TASK           — task instruction

    Parameters
    ----------
    slot : clean dict from extract_slot_data()
    """
    text        = slot["text"]
    bloom       = slot["bloom_level"]
    chunk_type  = slot["chunk_type"]
    concept     = slot["concept"] or "the main concept in this chunk"
    keywords    = slot["keywords"]
    prev_s      = slot["prev_sentence"]
    next_s      = slot["next_sentence"]

    bloom_instr = BLOOM_INSTRUCTION.get(bloom, BLOOM_INSTRUCTION["understand"])
    ct_instr    = CHUNK_TYPE_FORM.get(chunk_type,
                      "Use a general question style appropriate to the content.")
    kw_str      = ", ".join(keywords) if keywords else "(infer from text)"

    # Context fringe block (only included if present)
    context_lines = []
    if prev_s:
        context_lines.append(f"Previous: {prev_s}")
    if next_s:
        context_lines.append(f"Next    : {next_s}")

    parts = [
        "## CHUNK TEXT",
        text,
        "",
    ]

    if context_lines:
        parts += [
            "---",
            "",
            "## CONTEXT (background only — do NOT write questions about these sentences)",
        ] + context_lines + [""]

    parts += [
        "---",
        "",
        "## METADATA",
        f"- chunk_type : {chunk_type}",
        f"- keywords   : {kw_str}",
        "",
        "---",
        "",
        # ── Structured conditioning tags ──────────────────────────────────────
        f"[BLOOM LEVEL]: {bloom}",
        f"[CHUNK TYPE]: {chunk_type}",
        f"[FOCUS CONCEPT]: {concept}",
        "",
        "CHUNK TYPE INSTRUCTION",
        ct_instr,
        "",
        "---",
        "",
        "## YOUR TASK FOR THIS QUESTION",
        "Generate EXACTLY ONE MCQ. Not two. Not zero. One.",
        "Return a single JSON OBJECT (not an array).",
        f'The question MUST focus specifically on: "{concept}"',
        "The correct answer MUST be grounded in the chunk text above.",
    ]

    return "\n".join(parts)


print("User prompt builders defined: build_user_prompt_conditioned")
````

## 6. SFT Record Builders

`build_sft_record_conditioned()` and `build_sft_record_direct()` produce
LLaMA-Factory `sharegpt`-format records:

```json
{
  "id": "<unique record id>",
  "conversations": [
    {"role": "system",    "content": "<static system prompt>"},
    {"role": "user",      "content": "<chunk + metadata tags + task>"},
    {"role": "assistant", "content": "<validated MCQ as clean JSON string>"}
  ]
}
```

The assistant turn is a **clean JSON string** (not wrapped in markdown fences,
not an array) because:
- The system prompt explicitly instructs: *Return ONLY valid JSON. No markdown fences.*
- The student model must learn to produce raw JSON for its outputs to be parsed
  without post-processing at inference time.

````python
def _mcq_to_assistant_content(mcq: Dict) -> str:
    """
    Serialise the validated MCQ dict as a clean JSON string.

    - No markdown fences  (system prompt forbids them)
    - No array wrapper    (one slot = one MCQ object)
    - Compact but readable (indent=2 for human inspection; LLaMA-Factory ignores it)
    - Only the 4 MCQ fields: question, options, answer, explanation
    """
    return json.dumps(
        {
            "question":    mcq["question"],
            "options":     mcq["options"],
            "answer":      mcq["answer"],
            "explanation": mcq["explanation"],
        },
        ensure_ascii=False,
        indent=2,
    )


def build_sft_record_conditioned(slot: Dict) -> Dict:
    """
    Build a LLaMA-Factory sharegpt SFT record for OPTION A
    (metadata-conditioned generation).

    Record schema
    -------------
    {
      "id":            "<chunk_id>_slot<slot_index>_meta",
      "conversations": [
        {"role": "system",    "content": SFT_SYSTEM_PROMPT},
        {"role": "user",      "content": <conditioned user prompt>},
        {"role": "assistant", "content": <MCQ JSON string>}
      ]
    }

    Excluded fields: file_id, s5_validation, chunk_type_confidence,
                     stage_name, stage_number, timestamp, validation_result
    """
    return {
        "id": f"{slot['chunk_id']}_slot{slot['slot_index']}_meta",
        "conversations": [
            {
                "role":    "system",
                "content": SFT_SYSTEM_PROMPT,
            },
            {
                "role":    "user",
                "content": build_user_prompt_conditioned(slot),
            },
            {
                "role":    "assistant",
                "content": _mcq_to_assistant_content(slot["mcq"]),
            },
        ],
    }


print("SFT record builders defined: build_sft_record_conditioned")
````

## 7. Build SFT Dataset

Processes every record in `passed_mcqs.jsonl`:

1. Extract and validate slot data via `extract_slot_data()`
2. Deduplicate by `(chunk_id, slot_index)` — keeps the first occurrence
3. Build Option A records
4. Write both output files

````python
def build_sft_dataset(
    s5_records : List[Dict],
    s3_lookup  : Dict[str, Dict],
    meta_path  : str,
) -> Dict[str, int]:
    """
    Convert all S5 passed records into SFT training samples.

    Writes the Option A output file.

    Returns a stats dict.
    """

    stats = {
        "total_s5_records" : len(s5_records),
        "extracted"        : 0,
        "skipped_invalid"  : 0,
        "skipped_duplicate": 0,
        "written"          : 0,
    }

    seen_keys: set = set()   # (chunk_id, slot_index) — dedup guard

    with open(meta_path, "w", encoding="utf-8") as meta_f:

        # ── LOOP OVER ALL S5 RECORDS ───────────────────────────────
        for s5_rec in s5_records:

            # ── 1. Extract clean slot data ────────────────────────
            slot = extract_slot_data(s5_rec, s3_lookup)

            if slot is None:
                stats["skipped_invalid"] += 1
                continue

            stats["extracted"] += 1

            # ── 2. Deduplicate ───────────────────────────────────
            key = (slot["chunk_id"], slot["slot_index"])

            if key in seen_keys:
                stats["skipped_duplicate"] += 1
                continue

            seen_keys.add(key)

            # ── 3. Build SFT record ──────────────────────────────
            meta_rec = build_sft_record_conditioned(slot)

            meta_f.write(
                json.dumps(meta_rec, ensure_ascii=False) + "\n"
            )

            stats["written"] += 1

    return stats


# ── Run ───────────────────────────────────────────────────────────
stats = build_sft_dataset(
    s5_records = s5_records,
    s3_lookup  = s3_lookup,
    meta_path  = SFT_META_PATH
)

print("\n── SFT Dataset Build Complete ──────────────────────────")
print(f"  Total S5 records    : {stats['total_s5_records']:>6,}")
print(f"  Extracted (valid)   : {stats['extracted']:>6,}")
print(f"  Skipped (invalid)   : {stats['skipped_invalid']:>6,}")
print(f"  Skipped (duplicate) : {stats['skipped_duplicate']:>6,}")
print(f"  Written             : {stats['written']:>6,}")

print(f"\n  Output → {SFT_META_PATH}")
````

## 9. Inspection & Quality Checks

````python
# ── Reload and compute distribution statistics ────────────────────────────────
sft_records: List[Dict] = load_jsonl(SFT_META_PATH)

# Recover per-slot metadata from the clean id field
# (chunk_id and slot_index are embedded in the id, but we need to re-parse
#  slots from S5 for the stats — so we use the already-built slot list)

# Rebuild slot list for stats (does NOT re-write files)
slots_for_stats: List[Dict] = []
seen_stat_keys: set = set()
for s5_rec in s5_records:
    slot = extract_slot_data(s5_rec, s3_lookup)
    if slot is None:
        continue
    key = (slot["chunk_id"], slot["slot_index"])
    if key in seen_stat_keys:
        continue
    seen_stat_keys.add(key)
    slots_for_stats.append(slot)

bloom_dist  = Counter(s["bloom_level"] for s in slots_for_stats)
type_dist   = Counter(s["chunk_type"]  for s in slots_for_stats)
answer_dist = Counter(s["mcq"]["answer"] for s in slots_for_stats)
has_kw      = sum(1 for s in slots_for_stats if s["keywords"])
has_fringe  = sum(1 for s in slots_for_stats if s["prev_sentence"] or s["next_sentence"])

print(f"Total SFT records   : {len(sft_records):,}")
print()
print("── Bloom Level Distribution ────────────────────────────")
for level in ["remember", "understand", "apply", "analyze", "evaluate", "create"]:
    count = bloom_dist.get(level, 0)
    pct   = count / max(1, len(slots_for_stats)) * 100
    bar   = "█" * int(pct / 2)
    print(f"  {level:<12}: {count:>5}  ({pct:4.1f}%)  {bar}")

print()
print("── Chunk Type Distribution ─────────────────────────────")
for ctype in ["definition", "process", "comparison", "example", "argument"]:
    count = type_dist.get(ctype, 0)
    pct   = count / max(1, len(slots_for_stats)) * 100
    print(f"  {ctype:<12}: {count:>5}  ({pct:4.1f}%)")

print()
print("── Answer Key Distribution (should be ~25% each) ───────")
for key in ["A", "B", "C", "D"]:
    count = answer_dist.get(key, 0)
    pct   = count / max(1, len(slots_for_stats)) * 100
    bar   = "█" * int(pct / 2)
    print(f"  {key}: {count:>5}  ({pct:4.1f}%)  {bar}")

if answer_dist:
    max_share = max(answer_dist.values()) / max(1, len(slots_for_stats))
    flag = "✓ balanced" if max_share < 0.35 else "⚠ WARNING — positional bias detected"
    print(f"  Max share: {max_share:.1%}  {flag}")

print()
print("── Enrichment Coverage ─────────────────────────────────")
print(f"  Records with keywords      : {has_kw:>5} / {len(slots_for_stats)} ({has_kw/max(1,len(slots_for_stats)):.1%})")
print(f"  Records with context fringe: {has_fringe:>5} / {len(slots_for_stats)} ({has_fringe/max(1,len(slots_for_stats)):.1%})")
````

````python
# ── Schema validation: verify every record has the expected structure ──────────
errors: List[str] = []

for path, label in [(SFT_META_PATH, "Option A")]:
    records = load_jsonl(path)
    for i, rec in enumerate(records):
        rid = rec.get("id", f"record_{i}")

        # Top-level fields
        if "id" not in rec:
            errors.append(f"{label} record {i}: missing 'id'")
        if "conversations" not in rec or not isinstance(rec["conversations"], list):
            errors.append(f"{label} {rid}: missing or invalid 'conversations'")
            continue

        convs = rec["conversations"]
        if len(convs) != 3:
            errors.append(f"{label} {rid}: expected 3 turns, got {len(convs)}")
            continue

        roles = [c.get("role") for c in convs]
        if roles != ["system", "user", "assistant"]:
            errors.append(f"{label} {rid}: wrong roles {roles}")

        # Validate assistant turn is parseable JSON with required MCQ fields
        assistant_content = convs[2].get("content", "")
        try:
            mcq_parsed = json.loads(assistant_content)
            for field in ["question", "options", "answer", "explanation"]:
                if field not in mcq_parsed:
                    errors.append(f"{label} {rid}: assistant JSON missing '{field}'")
            if set(mcq_parsed.get("options", {}).keys()) != {"A", "B", "C", "D"}:
                errors.append(f"{label} {rid}: options keys are not exactly A,B,C,D")
            if mcq_parsed.get("answer") not in {"A", "B", "C", "D"}:
                errors.append(f"{label} {rid}: invalid answer key")
        except json.JSONDecodeError as exc:
            errors.append(f"{label} {rid}: assistant content is not valid JSON: {exc}")

        # Confirm excluded fields are absent from the top-level record
        for forbidden in ["file_id", "s5_validation", "chunk_type_confidence",
                          "stage_name", "stage_number", "timestamp", "validation_result"]:
            if forbidden in rec:
                errors.append(f"{label} {rid}: forbidden field '{forbidden}' present")

if errors:
    print(f"⚠  {len(errors)} validation error(s) found:")
    for e in errors[:20]:
        print(f"   {e}")
    if len(errors) > 20:
        print(f"   ... and {len(errors) - 20} more.")
else:
    print(f"✓  All records passed schema validation.")
    print(f"   Option A: {len(load_jsonl(SFT_META_PATH)):,} records — {SFT_META_PATH}")
````

### 11.1  Pre-flight Checks

Before writing any files we:
1. Verify the SFT JSONL exists and is non-empty
2. Compute token-length statistics → sets `CUTOFF_LEN` automatically
3. Audit Bloom-level distribution (needed for stratified split)
4. Audit answer-key distribution (detect positional bias A/B/C/D)


````python
import json, os, random
from collections import Counter, defaultdict
from os.path import join

# ── All paths come from Section 0 (Centralized Configuration) ────────────────
# No need to redefine LLAMA_FACTORY_DIR, INPUT_DIR, OUTPUT_DIR, SFT_META_PATH,
# FINETUNE_DIR, TRAIN_PATH, VAL_PATH — they are already set.

# Compatibility alias used by some cells below
FINETUNE_DIR = KAGGLE_FINETUNE_DIR

assert os.path.exists(SFT_META_PATH), (
    f"SFT file not found: {SFT_META_PATH}\n"
    f"Run sections 3–7 first to generate the SFT JSONL from the input dataset."
)

with open(SFT_META_PATH, encoding="utf-8") as f:
    ALL_RECORDS = [json.loads(l) for l in f if l.strip()]

print(f"Total SFT records loaded: {len(ALL_RECORDS):,}")
````

````python
# ── Token-length audit (4 chars ≈ 1 token heuristic) ─────────────────────────
def char_len(rec):
    return sum(len(t.get("content", "")) for t in rec.get("conversations", []))

tok_lens  = [char_len(r) // 4 for r in ALL_RECORDS]
sorted_tl = sorted(tok_lens)
n = len(sorted_tl)

p50 = sorted_tl[n // 2]
p90 = sorted_tl[int(0.90 * n)]
p95 = sorted_tl[int(0.95 * n)]
p99 = sorted_tl[int(0.99 * n)]

CUTOFF_LEN = min(4096, ((p95 // 512) + 1) * 512)
truncated  = sum(1 for t in tok_lens if t > CUTOFF_LEN)

print("Token length statistics (4-char heuristic):")
print(f"  Min  : {min(tok_lens):>6,}")
print(f"  P50  : {p50:>6,}")
print(f"  P90  : {p90:>6,}")
print(f"  P95  : {p95:>6,}")
print(f"  P99  : {p99:>6,}")
print(f"  Max  : {max(tok_lens):>6,}")
print(f"\nCUTOFF_LEN = {CUTOFF_LEN}  (P95 rounded to next 512, max 4096)")
print(f"Records that will be truncated: {truncated} ({truncated/n:.1%})")
````

````python
def get_bloom(rec):
    for turn in rec.get("conversations", []):
        if turn.get("role") == "user":
            for line in turn["content"].splitlines():
                if line.strip().startswith("[BLOOM LEVEL]"):
                    return line.split(":", 1)[-1].strip().lower()
    return "unknown"

bloom_labels = [get_bloom(r) for r in ALL_RECORDS]
bloom_dist   = Counter(bloom_labels)

print("Bloom level distribution:")
for lvl in ["remember","understand","apply","analyze","evaluate","create"]:
    c   = bloom_dist.get(lvl, 0)
    bar = "█" * int(c / max(bloom_dist.values(), default=1) * 30)
    print(f"  {lvl:<12}: {c:>5}  ({c/len(ALL_RECORDS):.1%})  {bar}")
````

````python
def get_answer(rec):
    for turn in rec.get("conversations", []):
        if turn.get("role") == "assistant":
            try:
                return json.loads(turn["content"]).get("answer", "?")
            except Exception:
                return "?"
    return "?"

answer_labels = [get_answer(r) for r in ALL_RECORDS]
ans_dist = Counter(answer_labels)

print("Answer distribution:")
for k in ["A","B","C","D"]:
    c = ans_dist.get(k, 0)
    print(f"  {k}: {c:>5}  ({c/len(ALL_RECORDS):.1%})")

max_share = max(ans_dist.get(k,0) for k in ["A","B","C","D"]) / len(ALL_RECORDS)
if max_share > 0.40:
    print("\n  WARNING: Positional bias detected (one answer key > 40%). Run debiasing cell below.")
else:
    print(f"\n  OK: Answer distribution balanced (max share {max_share:.1%}). No debiasing needed.")
````

````python
# ── (Optional) Answer-key debiasing ──────────────────────────────────────────
DEBIAS = False

if DEBIAS:
    by_answer = defaultdict(list)
    for rec, ans in zip(ALL_RECORDS, answer_labels):
        by_answer[ans].append(rec)
    cap = min(len(by_answer[k]) for k in ["A","B","C","D"]) * 2
    debiased = []
    rng = random.Random(42)
    for k in ["A","B","C","D"]:
        bucket = by_answer[k][:]
        rng.shuffle(bucket)
        debiased.extend(bucket[:cap])
    debiased.extend(by_answer.get("?", []))
    print(f"Before debiasing: {len(ALL_RECORDS):,}")
    ALL_RECORDS  = debiased
    bloom_labels = [get_bloom(r) for r in ALL_RECORDS]
    print(f"After debiasing : {len(ALL_RECORDS):,}")
else:
    print("Debiasing skipped.")
````

### 11.2  Create Train / Val / Test Splits

**Split ratios:** 85% train · 10% val · 5% test — **stratified by Bloom level**.


````python
# TRAIN_RATIO, VAL_RATIO, SPLIT_SEED come from Section 0.
SEED = SPLIT_SEED

by_bloom = defaultdict(list)
for rec, lvl in zip(ALL_RECORDS, bloom_labels):
    by_bloom[lvl].append(rec)

train_recs, val_recs = [], []
rng = random.Random(SEED)

for lvl, recs in by_bloom.items():
    rng.shuffle(recs)
    n = len(recs)
    n_val   = max(1, int(n * VAL_RATIO))
    n_train = n - n_val
    train_recs.extend(recs[:n_train])
    val_recs.extend(recs[n_train : n_train + n_val])

rng.shuffle(train_recs); rng.shuffle(val_recs)

total = len(train_recs) + len(val_recs)
assert total == len(ALL_RECORDS), "Split sum mismatch!"

print(f"Train : {len(train_recs):,}  ({len(train_recs)/len(ALL_RECORDS):.1%})")
print(f"Val   : {len(val_recs):,}  ({len(val_recs)/len(ALL_RECORDS):.1%})")
print(f"Total : {total:,}  OK")
````

````python
TRAIN_PATH = join(FINETUNE_DIR, "mcq_sft_train.jsonl")
VAL_PATH   = join(FINETUNE_DIR, "mcq_sft_val.jsonl")

def write_jsonl(path, records):
    with open(path, "w", encoding="utf-8") as f:
        for r in records:
            f.write(json.dumps(r, ensure_ascii=False) + "\n")
    print(f"  Written: {path}  ({len(records):,} records)")

write_jsonl(TRAIN_PATH, train_recs)
write_jsonl(VAL_PATH,   val_recs)
print("\nAll split files written.")
````

````python
import os, json, math
from os.path import join

# ── All directory and training constants come from Section 0 ──────────────────
# LLAMA_FACTORY_DIR, KAGGLE_LORA_DIR, TRAINING_YAML_PATH, RESUME_FROM_CHECKPOINT,
# NUM_ADDITIONAL_EPOCHS, LORA_RANK, LORA_ALPHA, LORA_DROPOUT, LEARNING_RATE,
# BATCH_SIZE, GRAD_ACCUMULATION, NEFTUNE_NOISE_ALPHA, WARMUP_RATIO,
# WEIGHT_DECAY, SAVE_TOTAL_LIMIT, RUN_NAME — all from centralized config.

WORLD_SIZE = int(os.environ.get("WORLD_SIZE", "1"))
cutoff = CUTOFF_LEN   # from 11.1 pre-flight checks

def _compute_epochs(ckpt_path, additional_epochs):
    # Read trainer_state.json to find epochs done, return total epochs for YAML
    if ckpt_path is None:
        return float(additional_epochs)
    state_path = os.path.join(ckpt_path, "trainer_state.json")
    if os.path.exists(state_path):
        try:
            with open(state_path) as f:
                state = json.load(f)
            done = float(state.get("epoch", 0.0))
            total = done + additional_epochs
            print(f"  Checkpoint epoch: {done:.2f} + {additional_epochs} more = {total:.2f} total")
            return total
        except Exception as exc:
            print(f"  Could not read trainer_state.json: {exc}")
    return float(additional_epochs)

_total_epochs = _compute_epochs(RESUME_FROM_CHECKPOINT, NUM_ADDITIONAL_EPOCHS)

# ── Checkpoint frequency: 2 per epoch ────────────────────────────────────────
try:
    _steps_per_epoch = math.ceil(
        len(train_recs) / (BATCH_SIZE * GRAD_ACCUMULATION * WORLD_SIZE)
    )
    _save_steps      = max(10, _steps_per_epoch // 2)
    _eval_steps      = _save_steps
    print(
        f"  Steps/epoch ~{_steps_per_epoch} (WORLD_SIZE={WORLD_SIZE}) "
        f"-> save/eval every {_save_steps} steps (~2 checkpoints/epoch)"
    )
except Exception:
    _save_steps = 82
    _eval_steps = 82
    print(f"  Fallback: save_steps={_save_steps}")

resume_line = (
    "resume_from_checkpoint: " + str(RESUME_FROM_CHECKPOINT)
    if RESUME_FROM_CHECKPOINT
    else "# resume_from_checkpoint: " + KAGGLE_LORA_DIR + "/checkpoint-NNN"
)

yaml_lines = [
    "",
    "### model",
    "model_name_or_path: " + BASE_MODEL_PATH,
    "trust_remote_code: true",
    "flash_attn: sdpa",
    "",
    "### method",
    "stage: sft",
    "do_train: true",
    "finetuning_type: lora",
    "",
    "### LoRA settings",
    "lora_rank: " + str(LORA_RANK),
    "lora_alpha: " + str(LORA_ALPHA),
    "lora_dropout: " + str(LORA_DROPOUT),
    "lora_target: all",
    "",
    "### dataset",
    "dataset: mcq_sft_train",
    "eval_dataset: mcq_sft_val",
    "template: qwen",
    "cutoff_len: " + str(cutoff),
    "train_on_prompt: false",
    "packing: false",
    "overwrite_cache: true",
    "preprocessing_num_workers: 8",
    "",
    "### output",
    "output_dir: " + KAGGLE_LORA_DIR,
    "logging_strategy: steps",
    "save_strategy: steps",
    "eval_strategy: steps",
    "logging_steps: 10",
    "save_steps: " + str(_save_steps),
    "eval_steps: " + str(_eval_steps),
    "save_total_limit: " + str(SAVE_TOTAL_LIMIT),
    "plot_loss: true",
    "include_num_input_tokens_seen: true",
    "",
    "### best-model tracking",
    "load_best_model_at_end: true",
    "metric_for_best_model: eval_loss",
    "greater_is_better: false",
    "",
    "### train",
    "per_device_train_batch_size: " + str(BATCH_SIZE),
    "gradient_accumulation_steps: " + str(GRAD_ACCUMULATION),
    "learning_rate: " + str(LEARNING_RATE),
    "num_train_epochs: " + str(round(_total_epochs, 1)),
    "lr_scheduler_type: cosine",
    "warmup_ratio: " + str(WARMUP_RATIO),
    "weight_decay: " + str(WEIGHT_DECAY),
    "bf16: true",
    "ddp_timeout: 180000000",
    "neftune_noise_alpha: " + str(NEFTUNE_NOISE_ALPHA),
    "",
    "### eval",
    "per_device_eval_batch_size: 1",
    "",
    "### Hugging Face Hub (disabled here — we upload manually in 11.8 with smart logic)",
    "push_to_hub: false",
    "export_hub_model_id: " + "LORA_REPO_ID_PLACEHOLDER",
    "hub_private_repo: true",
    "",
    "### tracking",
    'report_to: ["wandb"]',
    "run_name: " + RUN_NAME,
    "",
    "### resume",
    resume_line,
]

yaml_content = "\n".join(yaml_lines) + "\n"
# Substitute the placeholder after join (avoids f-string conflict)
yaml_content = yaml_content.replace("LORA_REPO_ID_PLACEHOLDER", LORA_REPO_ID)

os.makedirs(os.path.dirname(TRAINING_YAML_PATH), exist_ok=True)
with open(TRAINING_YAML_PATH, "w") as f:
    f.write(yaml_content)

print(f"\nTraining YAML written -> {TRAINING_YAML_PATH}")
print(f"  Model path        : {BASE_MODEL_PATH}")
print(f"  Epochs this run   : {_total_epochs:.1f}")
print(f"  Resume checkpoint : {RESUME_FROM_CHECKPOINT or 'None (fresh start)'}")
print()
print(yaml_content)
````

### 11.5  Pretty Training Progress Callback

A custom `TrainerCallback` that prints a structured, human-readable progress line
after every logging step and evaluation — visible in Kaggle's notebook output.

```
Step  120/500 | Epoch 0.75 | TrainLoss 1.245 | EvalLoss 1.102 | LR 2.1e-5 | GradNorm 0.84 | ETA 00:42:11
```


````python
import time, sys, re
from transformers import TrainerCallback

# ── ANSI escape stripper (tqdm may embed cursor codes in pty output) ──────────
_ANSI_RE = re.compile(r"\x1b\[[0-9;]*[A-Za-z]|\x1b\][^\x07]*\x07")

def _strip_ansi(s: str) -> str:
    return _ANSI_RE.sub("", s)

class PrettyProgressCallback(TrainerCallback):
    """
    Prints a live progress bar + structured metric line after every log/eval event.
    Works in-process (no subprocess), so the bar is always visible.
    """
    def __init__(self):
        self._train_start    = None
        self._best_eval_loss = float("inf")
        self._best_step      = None
        self._max_steps      = 1

    # ── helpers ───────────────────────────────────────────────────────────────
    def _bar(self, step: int, total: int, elapsed: float) -> str:
        pct    = int(step / max(total, 1) * 100)
        filled = int(30 * step / max(total, 1))
        bar    = "█" * filled + "░" * (30 - filled)
        eta_s  = (elapsed / max(step, 1)) * max(total - step, 0)
        speed  = step / max(elapsed, 1e-6)
        return (
            f"[{self._fmt(elapsed)}] {pct:>3}% |{bar}| {step}/{total} "
            f"[{self._fmt(elapsed)}<{self._fmt(eta_s)}, {speed:.2f}it/s]"
        )

    @staticmethod
    def _fmt(secs: float) -> str:
        secs = int(secs)
        return f"{secs // 3600:02d}:{(secs % 3600) // 60:02d}:{secs % 60:02d}"

    # ── Trainer hooks ─────────────────────────────────────────────────────────
    def on_train_begin(self, args, state, control, **kwargs):
        self._train_start = time.time()
        self._max_steps   = state.max_steps or 1
        print("=" * 80)
        print(f"  TRAINING START  |  Total steps: {state.max_steps}  |  Epochs: {args.num_train_epochs}")
        print("=" * 80, flush=True)

    def on_log(self, args, state, control, logs=None, **kwargs):
        if logs is None:
            return
        step    = state.global_step
        total   = state.max_steps or self._max_steps
        elapsed = time.time() - (self._train_start or time.time())

        # ── Progress bar (overwrites current line) ────────────────────────────
        bar_line = self._bar(step, total, elapsed)
        print(f"\r{bar_line}", end="", flush=True)

        # ── Collect readable metric parts ─────────────────────────────────────
        train_loss = logs.get("loss") or logs.get("train_loss")
        eval_loss  = logs.get("eval_loss")
        lr         = logs.get("learning_rate")
        grad_norm  = logs.get("grad_norm")
        epoch      = state.epoch or 0.0

        parts = [f"\nStep {step:>5}/{total}", f"Epoch {epoch:.3f}"]
        if train_loss is not None:
            parts.append(f"TrainLoss {float(train_loss):.4f}")
        if eval_loss is not None:
            parts.append(f"EvalLoss {float(eval_loss):.4f}")
            if float(eval_loss) < self._best_eval_loss:
                self._best_eval_loss = float(eval_loss)
                self._best_step      = step
                parts.append("⭐ NEW BEST")
        if lr        is not None: parts.append(f"LR {float(lr):.2e}")
        if grad_norm is not None: parts.append(f"GradNorm {float(grad_norm):.3f}")

        eta_s = (elapsed / max(step, 1)) * max(total - step, 0)
        parts.append(f"ETA {self._fmt(eta_s)}")
        print(" | ".join(parts), flush=True)

    def on_evaluate(self, args, state, control, metrics=None, **kwargs):
        if metrics and "eval_loss" in metrics:
            step = state.global_step
            el   = metrics["eval_loss"]
            tag  = " ⭐ NEW BEST" if el <= self._best_eval_loss else ""
            print(f"  📊 Eval @ step {step}  eval_loss={el:.4f}{tag}", flush=True)

    def on_save(self, args, state, control, **kwargs):
        step = state.global_step
        print(f"  💾 CHECKPOINT  Saving → {args.output_dir}/checkpoint-{step}", flush=True)
        if self._best_step == step:
            print(f"     ⭐ Best so far  eval_loss={self._best_eval_loss:.4f}", flush=True)

    def on_train_end(self, args, state, control, **kwargs):
        elapsed = time.time() - (self._train_start or time.time())
        total   = state.max_steps or self._max_steps
        print(f"\r{self._bar(total, total, elapsed)}", flush=True)
        print()
        print("=" * 80)
        print(f"  TRAINING COMPLETE  |  Total time: {self._fmt(elapsed)}")
        if self._best_step:
            print(f"  Best checkpoint: step {self._best_step}  |  eval_loss={self._best_eval_loss:.4f}")
        print("=" * 80, flush=True)

print("✓ PrettyProgressCallback defined.")
````

````python
# ── Model path verification (copy now done in cell 6) ────────────────────────
# Cell 6 handles the copy-and-rename. This cell just confirms BASE_MODEL_PATH
# is correct before training begins.

import os

assert os.path.exists(BASE_MODEL_PATH), f"BASE_MODEL_PATH not found: {BASE_MODEL_PATH}"
assert os.path.exists(os.path.join(BASE_MODEL_PATH, "model.safetensors")), (
    f"model.safetensors missing in {BASE_MODEL_PATH}"
)
assert os.path.exists(os.path.join(BASE_MODEL_PATH, "config.json")), (
    f"config.json missing in {BASE_MODEL_PATH}"
)

print(f"✓ BASE_MODEL_PATH = {BASE_MODEL_PATH}")
print(f"  Files: {sorted(os.listdir(BASE_MODEL_PATH))}")
````

````python
import os
import subprocess
import sys
import time
import re
import pty
import select
import errno

# LLAMA_FACTORY_DIR, TRAINING_YAML_PATH, WANDB_PROJECT_NAME come from Section 0.
YAML_PATH = TRAINING_YAML_PATH

os.environ["WANDB_PROJECT"]          = WANDB_PROJECT_NAME
os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["TQDM_DISABLE"]           = "0"
os.environ["TQDM_NCOLS"]             = "120"
os.environ["PYTHONUNBUFFERED"]       = "1"
# Suppress HuggingFace verbose logging at the library level
os.environ["TRANSFORMERS_VERBOSITY"] = "error"
os.environ["DATASETS_VERBOSITY"]     = "error"

# ── ANSI / control-character stripper ────────────────────────────────────────
_ANSI_RE = re.compile(r"\x1b\[[0-9;]*[A-Za-z]|\x1b\][^\x07]*\x07|\x1b[^\[\]][^\x1b]*")

def _strip_ansi(s: str) -> str:
    return _ANSI_RE.sub("", s)

# ── Noise filter: lines to suppress entirely ──────────────────────────────────
# Order matters: the INFO pattern catches ALL [INFO|...] lines first so we never
# accidentally pass raw HuggingFace trainer log lines to the metric formatter.
_HIDE_EXACT = [
    # HuggingFace / LLaMA-Factory INFO log lines  e.g. "[INFO|trainer.py:2385]"
    re.compile(r"\[INFO\|"),
    re.compile(r"\[WARNING\|"),
    re.compile(r"\[DEBUG\|"),
]
_HIDE_KEYWORDS = [
    "futurewarning", "userwarning", "nccl", "some weights of",
    "converting", "setting `pad_token`", "overflowing token",
    "tokenization", "deprecated", "cuda_fft", "cuda_dnn", "cuda_blas",
    "computation_placer", "absl::", "xla/stream", "warning:", "num_epochs",
    "num examples", "instantiating", "loading tokenizer", "loading model",
    "loading dataset", "special tokens", "special_tokens", "trainable params",
    "all params", "trainable%", "gradient checkpointing", "using", "initializing",
    "the following columns", "you are adding", "tokenizer class",
]

def _is_noise(line: str) -> bool:
    # Fast exact-RE check first (catches all [INFO|...] lines)
    for pat in _HIDE_EXACT:
        if pat.search(line):
            return True
    low = line.lower()
    return any(kw in low for kw in _HIDE_KEYWORDS)

# ── Progress-bar detector ─────────────────────────────────────────────────────
_BAR_RE = re.compile(r"\d+%\|")

def _is_progress_bar(line: str) -> bool:
    return bool(_BAR_RE.search(line))

# ── Metric dict detector and pretty-printer ───────────────────────────────────
# Matches raw trainer metric dicts like:  {'loss': 1.234, 'epoch': 0.5, ...}
_METRIC_DICT_RE = re.compile(
    r"\{[^}]*(?:loss|eval_loss|epoch|learning_rate|grad_norm|train_runtime)[^}]*\}",
    re.IGNORECASE,
)

def _fmt_metric_dict(line: str) -> str:
    """Extract and pretty-print metric dicts from a raw line."""
    m = _METRIC_DICT_RE.search(line)
    if not m:
        return None
    raw = m.group(0)
    try:
        # Replace single-quoted keys/values with double-quotes for json.loads
        import json as _json
        cleaned = raw.replace("'", '"')
        data = _json.loads(cleaned)
        parts = []
        if "loss" in data:        parts.append(f"loss={float(data['loss']):.4f}")
        if "eval_loss" in data:   parts.append(f"eval_loss={float(data['eval_loss']):.4f}")
        if "epoch" in data:       parts.append(f"epoch={float(data['epoch']):.3f}")
        if "learning_rate" in data: parts.append(f"lr={float(data['learning_rate']):.2e}")
        if "grad_norm" in data:   parts.append(f"grad_norm={float(data['grad_norm']):.3f}")
        if "train_runtime" in data:
            parts.append(f"runtime={float(data['train_runtime']):.1f}s")
        if "train_samples_per_second" in data:
            parts.append(f"samples/s={float(data['train_samples_per_second']):.2f}")
        return "  " + "  |  ".join(parts) if parts else None
    except Exception:
        return None

def _classify(line: str) -> str:
    """Return label: 'bar' | 'checkpoint' | 'best' | 'error' | 'metric' | 'skip'"""
    if _is_noise(line):
        return "skip"
    if _is_progress_bar(line):
        return "bar"
    low = line.lower()
    if "saving model checkpoint" in low or "checkpoint saved" in low:
        return "checkpoint"
    if "best model" in low or "loading best" in low:
        return "best"
    if "error" in low or "traceback" in low or "exception" in low or "oserror" in low:
        return "error"
    # Detect raw metric dicts OR structured metric lines
    _METRIC_KW = [
        "loss", "eval_loss", "epoch", "train_runtime",
        "train_samples_per_second", "learning_rate", "grad_norm",
    ]
    if any(kw in low for kw in _METRIC_KW):
        return "metric"
    return "skip"

# ── ETA tracker: parses tqdm lines into clean formatted output ────────────────
class _ETATracker:
    """
    Parses tqdm lines like ' 42%|████      | 61/145 [01:52<02:35,  1.85s/it]'
    and produces:  [00:01:52]  42% |████████████░░░░░░░░░░░░░░░░░░| 61/145 [01:52<02:35, 1.85s/it]
    """
    _RE = re.compile(
        r"(\d+)%\|.*?\|\s*(\d+)/(\d+)\s*\[(\d+:\d+)<(\d+:\d+|\?),\s*([\d.]+)([a-z/]+)\]",
        re.IGNORECASE,
    )

    def format(self, line: str, ts: str) -> str:
        m = self._RE.search(line)
        if not m:
            # Fallback: return cleaned line
            return f"{ts} {line.strip()}"
        pct, done, total, elapsed, eta, speed, unit = m.groups()
        done, total = int(done), int(total)
        bar_width   = 30
        filled      = int(bar_width * int(pct) / 100)
        bar         = "█" * filled + "░" * (bar_width - filled)
        return (
            f"{ts} {int(pct):>3}% |{bar}| {done}/{total} "
            f"[{elapsed}<{eta}, {speed}{unit}]"
        )

_eta         = _ETATracker()
_last_bar_line = ""   # track so we know when to emit a newline before non-bar output

# ── Startup banner ────────────────────────────────────────────────────────────
print("=" * 80)
print("  FINE-TUNING START")
print(f"  YAML   : {YAML_PATH}")
print(f"  Model  : {BASE_MODEL_PATH}")
print(f"  Epochs : {_total_epochs:.1f}  |  Resume: {RESUME_FROM_CHECKPOINT or 'fresh start'}")
print("=" * 80)
print()

t_start = time.time()

def _ts() -> str:
    e = time.time() - t_start
    return f"[{int(e // 3600):02d}:{int((e % 3600) // 60):02d}:{int(e % 60):02d}]"

def _flush_bar_newline():
    """If a bar line was the last output, start a new line before printing text."""
    global _last_bar_line
    if _last_bar_line:
        print()
        _last_bar_line = ""

def _emit(label: str, line: str) -> None:
    """Dispatch a classified line to the appropriate output format."""
    global _last_bar_line
    ts = _ts()

    if label == "bar":
        formatted = _eta.format(line, ts)
        print(f"\r{formatted}", end="", flush=True)
        _last_bar_line = formatted

    elif label == "checkpoint":
        _flush_bar_newline()
        print(f"\n  CHECKPOINT {ts} {line.strip()}", flush=True)

    elif label == "best":
        _flush_bar_newline()
        print(f"  BEST       {ts} {line.strip()}", flush=True)

    elif label == "error":
        _flush_bar_newline()
        print(f"  ERROR      {ts} {line.strip()}", flush=True)

    elif label == "metric":
        pretty = _fmt_metric_dict(line)
        if pretty:
            _flush_bar_newline()
            print(f"  {ts} {pretty}", flush=True)
        # If it's a plain metric line (not a raw dict), print it too
        elif not _METRIC_DICT_RE.search(line):
            _flush_bar_newline()
            print(f"  {ts} {line.strip()}", flush=True)
        # else: it was a raw dict but failed parsing — skip it silently

# ── Start subprocess in a pseudo-terminal (pty) so tqdm shows colours ────────
master_fd, slave_fd = pty.openpty()

proc = subprocess.Popen(
    ["llamafactory-cli", "train", YAML_PATH],
    cwd=LLAMA_FACTORY_DIR,
    env={**os.environ},
    stdin=slave_fd,
    stdout=slave_fd,
    stderr=slave_fd,
    close_fds=True,
)

os.close(slave_fd)

buffer = ""

try:
    while True:
        rlist, _, _ = select.select([master_fd], [], [], 0.5)

        if master_fd in rlist:
            try:
                chunk = os.read(master_fd, 4096)
            except OSError as exc:
                if exc.errno == errno.EIO:
                    break
                raise

            if not chunk:
                break

            # Decode and strip ANSI control codes before any processing
            buffer += _strip_ansi(chunk.decode("utf-8", errors="replace"))

            # Split on CR and LF; keep last partial fragment in buffer
            parts  = re.split(r"[\r\n]", buffer)
            buffer = parts.pop()

            for line in parts:
                line = line.rstrip()
                if not line:
                    continue
                label = _classify(line)
                if label != "skip":
                    _emit(label, line)

        if proc.poll() is not None and not rlist:
            break

    # Flush any remaining buffered text
    if buffer.strip():
        for line in re.split(r"[\r\n]", buffer):
            line = line.strip()
            if not line:
                continue
            label = _classify(line)
            if label != "skip":
                _emit(label, line)

finally:
    try:
        os.close(master_fd)
    except Exception:
        pass

if _last_bar_line:
    print()

proc.wait()

elapsed = time.time() - t_start
print()
print("=" * 80)
if proc.returncode == 0:
    print(
        f"  TRAINING COMPLETE  |  Time: "
        f"{int(elapsed // 3600):02d}:{int((elapsed % 3600) // 60):02d}:{int(elapsed % 60):02d}"
    )
else:
    print(f"  Training exited with code {proc.returncode}")
    print("  Check the ERROR lines above for details.")
print("=" * 80)
````

### 11.7  Best Checkpoint Selection

`load_best_model_at_end: true` in the YAML automatically restores the best eval-loss checkpoint.
This cell additionally scans checkpoints for JSON parse rate (a task-specific metric that
complements eval loss) and identifies the overall best checkpoint for export.


````python
import glob, os, json

# KAGGLE_LORA_DIR comes from Section 0.

# ── Read trainer_state.json to find best checkpoint by eval_loss ──────────────
trainer_state_path = os.path.join(KAGGLE_LORA_DIR, "trainer_state.json")

BEST_CKPT = None
best_eval_loss = float("inf")

if os.path.exists(trainer_state_path):
    with open(trainer_state_path) as f:
        state = json.load(f)

    print("Checkpoint history (eval_loss):")
    print(f"  {'Step':>6}  {'Eval Loss':>10}  {'Best?':>6}")
    print("  " + "-" * 30)

    for entry in state.get("log_history", []):
        if "eval_loss" in entry:
            step      = entry["step"]
            eval_loss = entry["eval_loss"]
            is_best   = ""
            if eval_loss < best_eval_loss:
                best_eval_loss = eval_loss
                is_best = "⭐"
                BEST_CKPT = os.path.join(KAGGLE_LORA_DIR, f"checkpoint-{step}")
            print(f"  {step:>6}  {eval_loss:>10.4f}  {is_best:>6}")

    if state.get("best_model_checkpoint"):
        BEST_CKPT = state["best_model_checkpoint"]
        print(f"\n  best_model_checkpoint from trainer_state: {BEST_CKPT}")

if BEST_CKPT and os.path.exists(BEST_CKPT):
    print(f"\n✓ Best checkpoint: {BEST_CKPT}  (eval_loss={best_eval_loss:.4f})")
else:
    BEST_CKPT = KAGGLE_LORA_DIR
    print(f"\n  Using full adapter dir as best checkpoint: {BEST_CKPT}")
````

### 11.8  Push All Checkpoints to Hugging Face Hub

Since `hub_strategy: all_checkpoints` is set in the YAML, LLaMA-Factory automatically pushes
every saved checkpoint to HF during training. This cell verifies and manually syncs any
checkpoints that may have been missed (e.g., after a session restart).

**Storage strategy:**
- ✅ Checkpoints → HF Hub (`CKPT_REPO_ID`) automatically during training
- ✅ Final best LoRA → HF Hub (`LORA_REPO_ID`) + `/kaggle/working/mcq-lora-adapter-best/`
- ✅ Final merged model → HF Hub (`MERGED_REPO_ID`) + `/kaggle/working/mcq-qwen-merged-best/`
- ❌ Checkpoints NOT kept permanently on Kaggle disk (deleted after HF push)


````python
# ── 11.8  Smart Checkpoint Upload ────────────────────────────────────────────
# All paths (KAGGLE_LORA_DIR, HF_BEST_TRACKER, CKPT_REPO_ID) from Section 0.

import glob, os, json, shutil
from huggingface_hub import HfApi

api = HfApi()

# ── Get eval_loss per step from trainer_state.json ────────────────────────────
trainer_state_path = os.path.join(KAGGLE_LORA_DIR, "trainer_state.json")
ckpt_eval_loss = {}
if os.path.exists(trainer_state_path):
    with open(trainer_state_path) as f:
        state = json.load(f)
    for entry in state.get("log_history", []):
        if "eval_loss" in entry:
            ckpt_eval_loss[entry["step"]] = entry["eval_loss"]
    print(f"Loaded eval_loss for {len(ckpt_eval_loss)} steps from trainer_state.json")
else:
    print("trainer_state.json not found — all checkpoints will be uploaded.")

def _load_hf_best():
    if os.path.exists(HF_BEST_TRACKER):
        try:
            with open(HF_BEST_TRACKER) as f:
                return json.load(f)
        except Exception:
            pass
    return {"best_eval_loss": float("inf"), "best_step": None}

def _save_hf_best(info):
    with open(HF_BEST_TRACKER, "w") as f:
        json.dump(info, f, indent=2)

hf_best = _load_hf_best()
print(f"Previously recorded HF best: eval_loss={hf_best['best_eval_loss']:.4f}  step={hf_best['best_step']}")

checkpoints = sorted(
    glob.glob(os.path.join(KAGGLE_LORA_DIR, "checkpoint-*")),
    key=lambda p: int(p.split("-")[-1])
)
print(f"\nFound {len(checkpoints)} local checkpoint(s).\n")

uploaded, skipped_worse, failed = 0, 0, 0

for ckpt in checkpoints:
    step = int(ckpt.split("-")[-1])
    el   = ckpt_eval_loss.get(step, None)
    folder_name = f"checkpoint-{step}"

    if el is not None and el >= hf_best["best_eval_loss"]:
        print(f"  SKIP   {folder_name}  eval_loss={el:.4f}  (>= HF best {hf_best['best_eval_loss']:.4f})")
        skipped_worse += 1
        continue

    el_str = f"{el:.4f}" if el is not None else "no-loss-data"
    print(f"  UPLOAD {folder_name}  eval_loss={el_str}  ...", end=" ", flush=True)
    try:
        api.upload_folder(
            folder_path=ckpt,
            repo_id=CKPT_REPO_ID,
            repo_type="model",
            path_in_repo=folder_name,
            token=HF_TOKEN,
        )
        print("ok")
        uploaded += 1
        if el is not None and el < hf_best["best_eval_loss"]:
            hf_best["best_eval_loss"] = el
            hf_best["best_step"]      = step
            _save_hf_best(hf_best)
            print(f"  -> New tracked best: step={step}, eval_loss={el:.4f}")
    except Exception as exc:
        print(f"FAILED: {exc}")
        failed += 1

print(f"\nUpload summary: {uploaded} uploaded, {skipped_worse} skipped (not better), {failed} failed.")

for ckpt in checkpoints:
    if ckpt != BEST_CKPT:
        shutil.rmtree(ckpt, ignore_errors=True)
        print(f"  Deleted local: {ckpt}")

print(f"\nCheckpoint sync done.  View at: https://huggingface.co/{CKPT_REPO_ID}")
````

### 11.15  Inference Helper

````python
import json as _json

def generate_mcq(messages: list, max_new_tokens: int = 512) -> dict:
    """
    Run MCQ inference with the merged fine-tuned model.

    Parameters
    ----------
    messages : list[dict]  [{role, content}, ...]  (system + user turns)
    max_new_tokens : int

    Returns
    -------
    dict  parsed MCQ {question, options, answer, explanation}
          or raw str if JSON parsing fails.
    """
    text = tokenizer.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )
    inputs = tokenizer([text], return_tensors="pt").to(DEVICE)

    with torch.no_grad():
        out_ids = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            temperature=None,
            top_p=None,
        )

    new_ids = out_ids[0][inputs.input_ids.shape[1]:]
    raw     = tokenizer.decode(new_ids, skip_special_tokens=True)

    try:
        clean = raw.strip().lstrip("```json").lstrip("```").rstrip("```").strip()
        return _json.loads(clean)
    except _json.JSONDecodeError:
        print("[WARN] Response is not valid JSON — returning raw string.")
        return raw

print("✓ generate_mcq() defined.")
````

## 12. LLM-as-a-Judge Evaluation Framework

This section evaluates three model stages on the **same** inputs using a reference-free LLM-as-a-Judge approach:

| Stage | Model | Description |
|---|---|---|
| **Stage 1 — Teacher** | Gemini flash-2.5-lite | MCQs from `teacher_outputs.jsonl` — the quality ceiling |
| **Stage 2 — Student Pre-FT** | Base Qwen2.5-1.5B-Instruct | MCQs generated **before** fine-tuning |
| **Stage 3 — Student Post-FT** | Merged Qwen (LoRA) | MCQs generated **after** fine-tuning |

**Inputs per MCQ (no reference MCQs required):**
- Generated MCQ (question, options A–D, correct answer, explanation)
- MCQ metadata: Bloom level, chunk_type, concept / learning objective
- Source chunk text (ground truth for factual grounding)
- Chunk metadata: topic / domain from `selected_subset_v3.jsonl`

**Six evaluation dimensions (1–5 Likert each):**

| ID | Dimension | Weight |
|---|---|---|
| D1 | Factual Grounding | 25% |
| D2 | Question Clarity | 10% |

### 12.2  Shared Utilities

Low-level helpers shared by every evaluation cell downstream:

- **`load_jsonl_safe`** — reads a JSONL file, skipping blank/malformed lines; returns `[]` if the file does not exist (safe for resumable pipelines).
- **`save_jsonl`** — appends or writes JSONL records.
- **`call_groq_judge`** — single-call wrapper around the Groq OpenAI-compatible API with retry/back-off logic.
- **`parse_judge_response`** — extracts the structured JSON score block from the judge's free-text reply.


````python
# ─── 12.2  Shared Utilities ──────────────────────────────────────────────────
import json, os, time, re
from typing import Dict, List, Optional, Tuple

# ── JSON-L helpers ────────────────────────────────────────────────────────────

def load_jsonl_safe(path: str) -> List[Dict]:
    """Load a JSONL, skipping blank/malformed lines."""
    if not os.path.exists(path):
        return []
    records = []
    with open(path, encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            try:
                records.append(json.loads(line))
            except json.JSONDecodeError:
                pass
    return records


def append_jsonl(path: str, record: Dict) -> None:
    """Append one record to a JSONL file (atomic per-sample save)."""
    with open(path, "a", encoding="utf-8") as f:
        f.write(json.dumps(record, ensure_ascii=False) + "\n")


def load_done_ids(scores_path: str) -> set:
    """Return the set of chunk_id+slot_index keys already scored (for resume)."""
    done = set()
    for r in load_jsonl_safe(scores_path):
        key = f"{r.get('chunk_id','')}_{r.get('slot_index','')}"
        done.add(key)
    return done


def record_key(rec: Dict) -> str:
    return f"{rec.get('chunk_id','')}_{rec.get('slot_index','')}"

# ── Chunk lookup ──────────────────────────────────────────────────────────────

def build_chunk_lookup(chunks_path: str) -> Dict[str, Dict]:
    """Build chunk_id → chunk_record lookup from selected_subset_v3.jsonl."""
    lookup = {}
    for r in load_jsonl_safe(chunks_path):
        cid = r.get("chunk_id")
        if cid:
            lookup[cid] = r
    print(f"Chunk lookup built: {len(lookup):,} chunks from {os.path.basename(chunks_path)}")
    return lookup

# ══════════════════════════════════════════════════════════════════════════════
# ── Groq LLM Judge Caller ────────────────────────────────────────────────────
# ══════════════════════════════════════════════════════════════════════════════

def call_judge(
    prompt_messages: List[Dict],
    model: str = None,
    max_retries: int = 3,
    retry_delay: float = 5.0,
) -> Optional[str]:
    """
    Call the LLM judge via Groq's OpenAI-compatible API.

    Uses the GROQ_API_KEY Kaggle secret and GROQ_BASE_URL.
    The model can be any Groq model slug (e.g. 'llama-3.3-70b-versatile').
    Get your free API key at https://console.groq.com/keys

    Parameters
    ----------
    prompt_messages : list of {role, content} dicts (system + user)
    model           : Groq model slug. Defaults to global JUDGE_MODEL.
    max_retries     : number of retry attempts on failure
    retry_delay     : seconds between retries (with exponential back-off)

    Returns
    -------
    Raw text content of the response, or None on total failure.
    """
    try:
        import openai
    except ImportError:
        raise ImportError("Run: pip install -q openai")

    if model is None:
        model = JUDGE_MODEL

    # Build the Groq client (uses OpenAI-compatible API)
    api_key = os.environ.get("GROQ_API_KEY", "")
    if not api_key:
        raise ValueError(
            "GROQ_API_KEY not found. "
            "Add it as a Kaggle Secret: Notebook → Add-ons → Secrets → "
            "Name: GROQ_API_KEY, Value: your key from https://console.groq.com/keys"
        )

    client = openai.OpenAI(
        base_url=GROQ_BASE_URL,
        api_key=api_key,
    )

    for attempt in range(1, max_retries + 1):
        try:
            resp = client.chat.completions.create(
                model=model,
                messages=prompt_messages,
                temperature=0.1,
                max_tokens=1024,
                # Note: response_format={"type":"json_object"} may not be
                # supported by all OpenRouter models. We handle JSON parsing
                # in parse_judge_json() with fallback stripping.
            )
            content = resp.choices[0].message.content
            if content:
                return content
            print(f"  [Judge] Attempt {attempt}: empty response, retrying...")
        except Exception as exc:
            print(f"  [Judge] Attempt {attempt}/{max_retries} failed: {exc}")
            if attempt < max_retries:
                wait = retry_delay * (2 ** (attempt - 1))  # exponential back-off
                time.sleep(wait)
    return None


def parse_judge_json(raw: Optional[str]) -> Optional[Dict]:
    """Parse judge response JSON, stripping any markdown fences."""
    if not raw:
        return None
    try:
        clean = raw.strip()
        # Strip markdown code fences if present
        clean = re.sub(r"^```json\s*", "", clean)
        clean = re.sub(r"^```\s*", "", clean)
        clean = re.sub(r"\s*```$", "", clean)
        return json.loads(clean)
    except json.JSONDecodeError:
        # Try to find JSON object in the response
        match = re.search(r'\{.*\}', raw, re.DOTALL)
        if match:
            try:
                return json.loads(match.group())
            except json.JSONDecodeError:
                pass
        return None


def compute_cmqs(scores: Dict) -> float:
    """Compute weighted CMQS from dimension scores dict."""
    weights = {
        "D1_factual_grounding":      0.25,
        "D2_question_clarity":       0.10,
        "D3_correct_answer_validity":0.20,
        "D4_distractor_plausibility":0.20,
        "D5_bloom_alignment":        0.15,
        "D6_topic_lo_alignment":     0.10,
    }
    total, weight_sum = 0.0, 0.0
    for key, w in weights.items():
        val = scores.get(key, {})
        s = val.get("score") if isinstance(val, dict) else val
        if s is not None:
            total += float(s) * w
            weight_sum += w
    if weight_sum == 0:
        return 0.0
    return round(total / weight_sum, 4)


print("✓ Shared utilities defined (Groq backend).")
print(f"  Judge model  : {JUDGE_MODEL}")
print(f"  Base URL     : {GROQ_BASE_URL}")
````

### 12.3  LLM Judge Prompt Builder

Constructs the **system** and **user** prompts sent to the judge model for each MCQ.

The judge scores every MCQ on **6 dimensions** (1–5 scale each):

| Dimension | Abbreviation | What it measures |
|---|---|---|
| Factual Grounding | D1 | Is the question grounded in the source chunk? |
| Question Clarity | D2 | Is the question unambiguous and well-formed? |
| Correct Answer Validity | D3 | Is the keyed answer demonstrably correct? |
| Distractor Plausibility | D4 | Are wrong options plausible but clearly wrong? |
| Bloom Alignment | D5 | Does the question match the requested Bloom level? |
| Topic & LO Alignment | D6 | Does it target the stated concept and learning objective? |

The **Composite MCQ Quality Score (CMQS)** is the weighted average of D1–D6 and is the primary evaluation metric throughout section 12.


````python
# ─── 12.3  LLM Judge Prompt Builder ─────────────────────────────────────────

JUDGE_SYSTEM_PROMPT = """You are an expert educational assessment evaluator with deep knowledge of Bloom's Taxonomy and MCQ item analysis.
You will receive a multiple-choice question (MCQ), its metadata, and the source text chunk it was generated from.
The source chunk is the ONLY source of truth for factual correctness.
Evaluate the MCQ on exactly six dimensions using a strict 1-5 integer scale.
Output ONLY valid JSON. No preamble, explanation, or text outside the JSON object.""".strip()

JUDGE_USER_TEMPLATE = """=== SOURCE CHUNK ===
{source_text}

=== CHUNK METADATA ===
Chunk Type: {chunk_type}
Concept / Topic: {concept}

=== GENERATED MCQ ===
Question: {question}
A) {opt_a}
B) {opt_b}
C) {opt_c}
D) {opt_d}
Correct Answer: {correct_answer}

=== MCQ METADATA ===
Bloom Taxonomy Level: {bloom_level}
Focus Concept: {focus_concept}

=== EVALUATION TASK ===
Score each dimension from 1 (worst) to 5 (best) using ONLY the provided content above.

D1 - Factual Grounding (weight 25%):
  Is every statement in the question stem AND all four options directly supported by the source chunk?
  5=fully grounded, 4=one minor implicit inference, 3=one ungrounded claim, 2=multiple hallucinations, 1=largely hallucinated

D2 - Question Clarity (weight 10%):
  Is the question grammatically correct, unambiguous, and self-contained?
  5=perfect, 4=minor verbosity/grammar, 3=some ambiguity, 2=notably unclear, 1=incomprehensible

D3 - Correct Answer Validity (weight 20%):
  Is the designated correct answer factually correct per the source chunk AND unambiguously better than all distractors?
  5=unambiguously correct, 4=correct but slightly imprecise, 3=correct but a distractor is close, 2=partially correct, 1=wrong

D4 - Distractor Plausibility (weight 20%):
  Are ALL three distractors plausible to a student without mastery? Clearly wrong only to an expert?
  5=all highly plausible/misconception-based, 4=mostly plausible, 3=mixed quality, 2=mostly implausible, 1=trivially wrong/irrelevant

D5 - Bloom Alignment (weight 15%):
  Does the cognitive demand required to answer match the assigned Bloom level ({bloom_level})?
  Levels: remember<understand<apply<analyze<evaluate<create
  5=exact match, 4=one adjacent level, 3=adjacent level, 2=two+ levels off, 1=completely wrong

D6 - Topic & Learning Objective Alignment (weight 10%):
  Does the question genuinely test the focus concept and match the chunk's topic?
  5=fully aligned, 4=mostly aligned, 3=loosely related, 2=tangentially related, 1=off-topic

Return EXACTLY this JSON structure (integer scores, one-sentence reasons):
{{
  "D1_factual_grounding":       {{"score": <1-5>, "reason": "<one sentence>"}},
  "D2_question_clarity":        {{"score": <1-5>, "reason": "<one sentence>"}},
  "D3_correct_answer_validity": {{"score": <1-5>, "reason": "<one sentence>"}},
  "D4_distractor_plausibility": {{"score": <1-5>, "reason": "<one sentence>"}},
  "D5_bloom_alignment":         {{"score": <1-5>, "reason": "<one sentence>"}},
  "D6_topic_lo_alignment":      {{"score": <1-5>, "reason": "<one sentence>"}}
}}""".strip()


def build_judge_messages(mcq_record: Dict, chunk_record: Dict) -> List[Dict]:
    """
    Build the [system, user] message list for the LLM judge.

    Parameters
    ----------
    mcq_record   : one record from teacher / student output JSONL
                   Must have: chunk_id, text, bloom_level, chunk_type,
                              concept, mcq {question, options, answer}
    chunk_record : matching record from selected_subset_v3.jsonl
                   Must have: text, metadata {bloom_level, chunk_type, concepts}
    """
    # ── Extract MCQ fields ─────────────────────────────────────────────────────
    mcq      = mcq_record.get("mcq", {})
    question = mcq.get("question", "").strip()
    options  = mcq.get("options", {})
    answer   = mcq.get("answer", "").strip().upper()
    bloom    = mcq_record.get("bloom_level", "").strip()
    concept  = mcq_record.get("concept", "").strip()
    chunk_t  = mcq_record.get("chunk_type", "").strip()

    # Source text: prefer chunk_record text (canonical), fall back to mcq_record text
    source_text = chunk_record.get("text", mcq_record.get("text", "")).strip()

    # Chunk metadata for domain / concept context
    chunk_meta   = chunk_record.get("metadata", {})
    chunk_type_c = chunk_meta.get("chunk_type", chunk_t) or chunk_t
    concepts_c   = chunk_meta.get("concepts", [concept]) or [concept]
    concept_str  = ", ".join(concepts_c) if concepts_c else concept

    user_content = JUDGE_USER_TEMPLATE.format(
        source_text    = source_text,
        chunk_type     = chunk_type_c,
        concept        = concept_str,
        question       = question,
        opt_a          = options.get("A", ""),
        opt_b          = options.get("B", ""),
        opt_c          = options.get("C", ""),
        opt_d          = options.get("D", ""),
        correct_answer = answer,
        bloom_level    = bloom,
        focus_concept  = concept,
    )

    return [
        {"role": "system", "content": JUDGE_SYSTEM_PROMPT},
        {"role": "user",   "content": user_content},
    ]


print("✓ Judge prompt builder defined.")
print(f"  System prompt : {len(JUDGE_SYSTEM_PROMPT):,} chars")
print(f"  User template : {len(JUDGE_USER_TEMPLATE):,} chars")
````

### 12.4  Core Evaluation Runner (Resumable)

The generic runner that drives all three judge stages (teacher, pre-FT, post-FT).

**Key design decisions:**
- **Resumable** — already-scored records are loaded at start; only unscored items are sent to the judge, so a crashed run can be continued without re-spending API quota.
- **Rate-limit aware** — configurable `rpm_limit` with sleep between batches.
- **Idempotent** — safe to re-run; duplicate keys are deduplicated on load.

Outputs one JSONL line per MCQ with fields: `chunk_id`, `slot_index`, `bloom_level`, `CMQS`, `scores` (per-dimension dict), `status` (`ok` | `error`), `judge_model`.


````python
# ─── 12.4  Core Evaluation Runner (resumable) ────────────────────────────────
from tqdm.auto import tqdm

def evaluate_stage(
    records:        List[Dict],
    chunk_lookup:   Dict[str, Dict],
    scores_path:    str,
    stage_label:    str,
) -> List[Dict]:
    """
    Run LLM-as-a-Judge evaluation on a list of MCQ records.
    Fully resumable: already-scored records are skipped.

    Parameters
    ----------
    records       : list of MCQ records (teacher or student output JSONL)
    chunk_lookup  : chunk_id -> chunk_record from selected_subset_v3.jsonl
    scores_path   : path to JSONL file where scores are appended
    stage_label   : "teacher" | "student_pre_ft" | "student_post_ft"

    Returns
    -------
    list of all score records (previously saved + newly computed)
    """
    done_ids   = load_done_ids(scores_path)
    all_scores = load_jsonl_safe(scores_path)

    to_score = [r for r in records if record_key(r) not in done_ids]
    print(f"\n{'='*60}")
    print(f"Stage        : {stage_label}")
    print(f"Total records: {len(records):,}")
    print(f"Already done : {len(done_ids):,}")
    print(f"To evaluate  : {len(to_score):,}")
    print(f"Scores file  : {scores_path}")
    print(f"{'='*60}")

    if not to_score:
        print("✓ All records already evaluated. Nothing to do.")
        return all_scores

    skipped, failed = 0, 0

    for rec in tqdm(to_score, desc=f"Judging [{stage_label}]"):
        chunk_id    = rec.get("chunk_id", "")
        slot_index  = rec.get("slot_index", 0)
        chunk_rec   = chunk_lookup.get(chunk_id, {})

        # Skip records with no matching chunk (can't evaluate grounding)
        if not chunk_rec:
            skipped += 1
            score_rec = {
                "chunk_id":    chunk_id,
                "slot_index":  slot_index,
                "stage":       stage_label,
                "status":      "skipped_no_chunk",
                "CMQS":        None,
                "scores":      None,
            }
            append_jsonl(scores_path, score_rec)
            all_scores.append(score_rec)
            continue

        # Validate MCQ structure
        mcq = rec.get("mcq", {})
        if not mcq.get("question") or not mcq.get("options") or not mcq.get("answer"):
            skipped += 1
            score_rec = {
                "chunk_id":   chunk_id,
                "slot_index": slot_index,
                "stage":      stage_label,
                "status":     "skipped_invalid_mcq",
                "CMQS":       None,
                "scores":     None,
            }
            append_jsonl(scores_path, score_rec)
            all_scores.append(score_rec)
            continue

        # Build judge prompt and call judge
        messages = build_judge_messages(rec, chunk_rec)
        raw_resp = call_judge(messages, model=JUDGE_MODEL)
        parsed   = parse_judge_json(raw_resp)

        if parsed is None:
            failed += 1
            score_rec = {
                "chunk_id":    chunk_id,
                "slot_index":  slot_index,
                "stage":       stage_label,
                "status":      "judge_parse_failed",
                "raw_response":raw_resp,
                "CMQS":        None,
                "scores":      None,
            }
            append_jsonl(scores_path, score_rec)
            all_scores.append(score_rec)
            continue

        # Validate all 6 dimension scores are present and in [1,5]
        dim_keys = [
            "D1_factual_grounding", "D2_question_clarity",
            "D3_correct_answer_validity", "D4_distractor_plausibility",
            "D5_bloom_alignment", "D6_topic_lo_alignment",
        ]
        dim_scores_valid = all(
            isinstance(parsed.get(k, {}).get("score"), (int, float))
            and 1 <= parsed[k]["score"] <= 5
            for k in dim_keys
            if k in parsed
        )
        if not dim_scores_valid or not all(k in parsed for k in dim_keys):
            failed += 1
            score_rec = {
                "chunk_id":    chunk_id,
                "slot_index":  slot_index,
                "stage":       stage_label,
                "status":      "judge_schema_invalid",
                "raw_response":raw_resp,
                "CMQS":        None,
                "scores":      parsed,
            }
            append_jsonl(scores_path, score_rec)
            all_scores.append(score_rec)
            continue

        # Success — compute CMQS and save
        cmqs = compute_cmqs(parsed)
        score_rec = {
            "chunk_id":   chunk_id,
            "slot_index": slot_index,
            "bloom_level":rec.get("bloom_level", ""),
            "chunk_type": rec.get("chunk_type", ""),
            "stage":      stage_label,
            "status":     "ok",
            "CMQS":       cmqs,
            "scores": {
                k: parsed[k]
                for k in dim_keys
            },
        }
        append_jsonl(scores_path, score_rec)
        all_scores.append(score_rec)

    print(f"\n✓ Done.  Skipped={skipped}  Failed={failed}  "
          f"Scored={len(to_score)-skipped-failed}")
    return all_scores


print("✓ evaluate_stage() defined.")
````

### 12.5  Build Shared Evaluation Sample

Constructs the **shared evaluation set** used by all three stages.

Reads from `selected_subset_v3.jsonl` (S3 chunks) and samples up to `EVAL_SAMPLE_N` chunks (set in Section 0, default 200).
Using the **same chunks** across teacher, pre-FT, and post-FT ensures fair apples-to-apples comparison — every model is asked to generate an MCQ for the exact same source text.

Output: `eval_chunks` — a list of chunk dicts passed to each inference cell.


````python
# ─── 12.5  Build Shared Evaluation Sample ────────────────────────────────────
# We evaluate all three models on the SAME set of chunk inputs.
# The teacher JSONL is the reference for which chunks to evaluate.
# Student models are run on the same chunk_ids.

import random
from collections import defaultdict

# ── Load teacher outputs & chunk lookup ──────────────────────────────────────
chunk_lookup = build_chunk_lookup(CHUNKS_PATH)

teacher_records_all = load_jsonl_safe(TEACHER_OUTPUT_PATH)
if not teacher_records_all:
    raise FileNotFoundError(
        f"Teacher output file not found or empty: {TEACHER_OUTPUT_PATH}\n"
        "Ensure 'teacher_outputs.jsonl' is in your Kaggle input dataset."
    )
print(f"Teacher records loaded: {len(teacher_records_all):,}")

# ── Stratified sample by Bloom level (or use all) ────────────────────────────
def stratified_sample(records: List[Dict], n: Optional[int], seed: int = 42) -> List[Dict]:
    """
    Return a stratified random sample of n records balanced by Bloom level.
    If n is None or n >= len(records), returns all records (shuffled).
    """
    if n is None or n >= len(records):
        rng = random.Random(seed)
        out = records[:]
        rng.shuffle(out)
        return out

    by_bloom = defaultdict(list)
    for r in records:
        by_bloom[r.get("bloom_level", "unknown")].append(r)

    n_levels  = len(by_bloom)
    per_level = max(1, n // n_levels)
    rng       = random.Random(seed)
    sampled   = []
    for bloom_recs in by_bloom.values():
        rng.shuffle(bloom_recs)
        sampled.extend(bloom_recs[:per_level])

    # top-up if under n due to rounding
    rng.shuffle(sampled)
    remaining = [r for r in records if r not in sampled]
    rng.shuffle(remaining)
    sampled.extend(remaining[:max(0, n - len(sampled))])
    rng.shuffle(sampled)
    return sampled[:n]

teacher_eval_records = stratified_sample(teacher_records_all, EVAL_SAMPLE_N)
eval_chunk_ids       = {r["chunk_id"] for r in teacher_eval_records}

print(f"\nEvaluation sample:")
print(f"  Teacher records : {len(teacher_eval_records):,}")
print(f"  Unique chunk_ids: {len(eval_chunk_ids):,}")

from collections import Counter
bloom_dist = Counter(r.get("bloom_level", "?") for r in teacher_eval_records)
print("  Bloom distribution:")
for lvl, cnt in sorted(bloom_dist.items()):
    print(f"    {lvl:<12}: {cnt:3d}  ({cnt/len(teacher_eval_records):.1%})")
````

### 12.6  Student Model Inference — Pre-Fine-Tuning

Generates MCQs using the **base Qwen2.5-1.5B-Instruct** model (before fine-tuning).
Uses the exact same system + user prompt format as the SFT training data, so the comparison reflects what the model can do *without* distillation.

**Fully resumable** — already-generated records are loaded from `STUDENT_PRE_OUTPUT_PATH`; only missing slots are inferred.

Output: `student_pre_ft_outputs.jsonl`


````python
def _run_student_inference(
    model, tokenizer, messages: List[Dict], max_new_tokens: int = 512
) -> dict:
    """Run a single inference call. Returns parsed MCQ dict or raw str."""
    text   = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer([text], return_tensors="pt").to(DEVICE)
    with torch.no_grad():
        out_ids = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            temperature=None,
            top_p=None,
        )
    new_ids = out_ids[0][inputs.input_ids.shape[1]:]
    raw     = tokenizer.decode(new_ids, skip_special_tokens=True)
    try:
        clean = raw.strip()
        clean = re.sub(r"^```json\s*", "", clean)
        clean = re.sub(r"^```\s*", "", clean)
        clean = re.sub(r"\s*```$", "", clean)
        return json.loads(clean)
    except json.JSONDecodeError:
        return {"_raw": raw}


def generate_student_outputs(
    model, tokenizer,
    teacher_eval_records: List[Dict],
    chunk_lookup:         Dict[str, Dict],
    output_path:          str,
    stage_label:          str,
) -> List[Dict]:
    """
    Generate MCQ outputs for all teacher eval records using the given model.
    Uses the EXACT same SFT prompt format (SFT_SYSTEM_PROMPT + build_user_prompt_conditioned).
    Fully resumable: skips already-generated samples.
    Returns list of all output records.
    """
    # Load already-done IDs
    existing     = load_jsonl_safe(output_path)
    done_keys    = {record_key(r) for r in existing}
    all_outputs  = existing[:]
    to_generate  = [r for r in teacher_eval_records if record_key(r) not in done_keys]

    print(f"\n{'='*60}")
    print(f"Stage         : {stage_label}")
    print(f"Total records : {len(teacher_eval_records):,}")
    print(f"Already done  : {len(done_keys):,}")
    print(f"To generate   : {len(to_generate):,}")
    print(f"Output file   : {output_path}")
    print(f"{'='*60}")

    if not to_generate:
        print("✓ All outputs already generated.")
        return all_outputs

    for rec in tqdm(to_generate, desc=f"Generating [{stage_label}]"):
        chunk_id   = rec.get("chunk_id", "")
        slot_index = rec.get("slot_index", 0)
        chunk_rec  = chunk_lookup.get(chunk_id, {})

        # Build the slot dict expected by build_user_prompt_conditioned
        s3_meta   = chunk_rec.get("metadata", {})
        fringe    = chunk_rec.get("context_fringe", {})
        slot = {
            "text":          chunk_rec.get("text", rec.get("text", "")),
            "bloom_level":   rec.get("bloom_level", s3_meta.get("bloom_level", "understand")),
            "chunk_type":    rec.get("chunk_type",  s3_meta.get("chunk_type", "definition")),
            "concept":       rec.get("concept",     (s3_meta.get("concepts") or [""])[0]),
            "keywords":      s3_meta.get("keywords", []),
            "prev_sentence": (fringe.get("prev_sentence") or "").strip(),
            "next_sentence": (fringe.get("next_sentence") or "").strip(),
        }

        messages = [
            {"role": "system", "content": SFT_SYSTEM_PROMPT},
            {"role": "user",   "content": build_user_prompt_conditioned(slot)},
        ]

        generated_mcq = _run_student_inference(model, tokenizer, messages)

        output_rec = {
            "chunk_id":   chunk_id,
            "slot_index": slot_index,
            "text":       slot["text"],
            "chunk_type": slot["chunk_type"],
            "concept":    slot["concept"],
            "bloom_level":slot["bloom_level"],
            "stage":      stage_label,
            "mcq": generated_mcq if isinstance(generated_mcq, dict) and "_raw" not in generated_mcq
                   else {"question": "", "options": {}, "answer": "", "explanation": "",
                         "_raw_unparsed": generated_mcq.get("_raw", "")},
        }
        append_jsonl(output_path, output_rec)
        all_outputs.append(output_rec)

    print(f"\n✓ Generation complete: {len(all_outputs):,} total records in {output_path}")
    return all_outputs
````

### 12.9  Compute Stage Statistics & Quality Preservation

Aggregates raw judge scores into **summary statistics** for all three stages.

Computed per stage and per dimension:
- **Mean, Std, Median, Min, Max** of CMQS and each D1–D6 score
- **Bloom-level CMQS breakdown** — mean CMQS per Bloom taxonomy level (remember → create)
- **Quality label distribution** — percentage of MCQs in each bucket (Excellent ≥ 4.5 / Good ≥ 3.75 / Acceptable ≥ 3.0 / Poor ≥ 2.0 / Reject < 2.0)
- **Quality Preservation Score (QPS)** = Post-FT CMQS / Teacher CMQS × 100

All stats are saved to `final_comparison_results.json` (the master evaluation record).


````python
# ─── 12.9  Compute Stage Statistics & Quality Preservation ───────────────────
import statistics, json

DIM_KEYS = [
    "D1_factual_grounding",
    "D2_question_clarity",
    "D3_correct_answer_validity",
    "D4_distractor_plausibility",
    "D5_bloom_alignment",
    "D6_topic_lo_alignment",
]

DIM_LABELS = {
    "D1_factual_grounding":       "D1 — Factual Grounding",
    "D2_question_clarity":        "D2 — Question Clarity",
    "D3_correct_answer_validity": "D3 — Correct Answer Validity",
    "D4_distractor_plausibility": "D4 — Distractor Plausibility",
    "D5_bloom_alignment":         "D5 — Bloom Alignment",
    "D6_topic_lo_alignment":      "D6 — Topic & LO Alignment",
}

def compute_stage_stats(score_records: List[Dict]) -> Dict:
    """Compute per-dimension and CMQS mean/std/median for valid (status=ok) records."""
    valid = [r for r in score_records if r.get("status") == "ok"]
    if not valid:
        return {}

    stats = {}

    # CMQS overall
    cmqs_vals = [r["CMQS"] for r in valid if r.get("CMQS") is not None]
    stats["CMQS"] = {
        "mean":   round(statistics.mean(cmqs_vals), 4),
        "std":    round(statistics.stdev(cmqs_vals) if len(cmqs_vals) > 1 else 0.0, 4),
        "median": round(statistics.median(cmqs_vals), 4),
        "min":    round(min(cmqs_vals), 4),
        "max":    round(max(cmqs_vals), 4),
        "n":      len(cmqs_vals),
    }

    # Per dimension
    for key in DIM_KEYS:
        vals = []
        for r in valid:
            dim = (r.get("scores") or {}).get(key, {})
            s   = dim.get("score") if isinstance(dim, dict) else dim
            if s is not None:
                vals.append(float(s))
        if vals:
            stats[key] = {
                "mean":   round(statistics.mean(vals), 4),
                "std":    round(statistics.stdev(vals) if len(vals) > 1 else 0.0, 4),
                "median": round(statistics.median(vals), 4),
                "n":      len(vals),
            }

    # Bloom-level CMQS breakdown
    bloom_cmqs = {}
    for r in valid:
        bl  = r.get("bloom_level", "unknown")
        val = r.get("CMQS")
        if val is not None:
            bloom_cmqs.setdefault(bl, []).append(val)
    stats["bloom_breakdown"] = {
        bl: {
            "mean": round(statistics.mean(vs), 4),
            "std":  round(statistics.stdev(vs) if len(vs) > 1 else 0.0, 4),
            "n":    len(vs),
        }
        for bl, vs in sorted(bloom_cmqs.items())
    }

    # Quality label distribution
    label_counts = {"Excellent": 0, "Good": 0, "Acceptable": 0, "Poor": 0, "Reject": 0}
    for v in cmqs_vals:
        if   v >= 4.50: label_counts["Excellent"]  += 1
        elif v >= 3.75: label_counts["Good"]        += 1
        elif v >= 3.00: label_counts["Acceptable"]  += 1
        elif v >= 2.00: label_counts["Poor"]        += 1
        else:           label_counts["Reject"]      += 1
    stats["quality_distribution"] = {
        k: {"count": v, "pct": round(v / len(cmqs_vals) * 100, 1)}
        for k, v in label_counts.items()
    }

    stats["n_valid"] = len(valid)
    stats["n_total"] = len(score_records)
    return stats


# ── Compute stats for all three stages ───────────────────────────────────────
stats_teacher  = compute_stage_stats(teacher_scores)
stats_pre_ft   = compute_stage_stats(student_pre_scores)
stats_post_ft  = compute_stage_stats(student_post_scores)

# ── Quality Preservation Score ────────────────────────────────────────────────
def qps(post_val, teacher_val):
    if teacher_val and teacher_val > 0:
        return round(post_val / teacher_val * 100, 2)
    return None

teacher_cmqs  = stats_teacher.get("CMQS", {}).get("mean", 0)
post_ft_cmqs  = stats_post_ft.get("CMQS", {}).get("mean", 0)
pre_ft_cmqs   = stats_pre_ft.get("CMQS", {}).get("mean", 0)

overall_qps = qps(post_ft_cmqs, teacher_cmqs)
pre_ft_qps  = qps(pre_ft_cmqs, teacher_cmqs)
ft_gain     = round(post_ft_cmqs - pre_ft_cmqs, 4) if pre_ft_cmqs else None

print("\n" + "="*60)
print("EVALUATION RESULTS — SUMMARY")
print("="*60)
print(f"{'Metric':<35} {'Teacher':>9} {'Pre-FT':>9} {'Post-FT':>9}")
print("-"*62)
print(f"{'CMQS (mean ± std)':<35} "
      f"{stats_teacher.get('CMQS',{}).get('mean','-'):>6.3f} ± {stats_teacher.get('CMQS',{}).get('std',0):>4.3f}  "
      f"{stats_pre_ft.get('CMQS',{}).get('mean','-'):>6.3f} ± {stats_pre_ft.get('CMQS',{}).get('std',0):>4.3f}  "
      f"{stats_post_ft.get('CMQS',{}).get('mean','-'):>6.3f} ± {stats_post_ft.get('CMQS',{}).get('std',0):>4.3f}")
print()
for key in DIM_KEYS:
    t = stats_teacher.get(key, {}).get("mean", "-")
    b = stats_pre_ft.get(key, {}).get("mean", "-")
    a = stats_post_ft.get(key, {}).get("mean", "-")
    label = DIM_LABELS[key]
    print(f"  {label:<33} {t:>9.3f}  {b:>9.3f}  {a:>9.3f}")
print()
print(f"Quality Preservation Score (QPS):   {overall_qps}%")
print(f"Fine-Tuning Gain (Post - Pre CMQS): {ft_gain:+.4f}" if ft_gain is not None else "N/A")
print("="*60)

# ── Save all stats ────────────────────────────────────────────────────────────
all_stats = {
    "teacher":         stats_teacher,
    "student_pre_ft":  stats_pre_ft,
    "student_post_ft": stats_post_ft,
    "summary": {
        "QPS_overall_pct":   overall_qps,
        "FT_gain_CMQS":      ft_gain,
        "teacher_CMQS_mean": teacher_cmqs,
        "pre_ft_CMQS_mean":  pre_ft_cmqs,
        "post_ft_CMQS_mean": post_ft_cmqs,
    }
}
with open(COMPARISON_PATH, "w", encoding="utf-8") as f:
    json.dump(all_stats, f, indent=2, ensure_ascii=False)
print(f"\n✓ Full comparison results saved → {COMPARISON_PATH}")
````

### 12.11  P1 — Statistical Significance Tests (Wilcoxon + Cohen's d)

Tests whether the Pre-FT → Post-FT improvement and the Teacher → Post-FT gap are **statistically significant**.

**Method:** Wilcoxon signed-rank test (non-parametric; appropriate for ordinal 1–5 Likert scores on paired samples matched by `chunk_id + slot_index`). Reports one-sided p-values (alternative = `greater`) and Cohen's d effect size.

| Test | Question answered |
|---|---|
| Post-FT > Pre-FT (CMQS) | Did fine-tuning significantly improve quality? |
| Teacher > Post-FT (CMQS) | Is the residual teacher–student gap significant? |
| Per-dimension Post-FT > Pre-FT | Which dimensions improved significantly? |

**Effect size interpretation:** negligible < 0.2 · small < 0.5 · medium < 0.8 · large ≥ 0.8

Results are appended to `final_comparison_results.json` under `significance_tests`.


````python
# ─── 12.11  P1 — Wilcoxon Signed-Rank Test + Cohen's d ──────────────────────
# Tests whether Pre-FT→Post-FT and Teacher→Post-FT differences are statistically
# significant. Uses Wilcoxon signed-rank (non-parametric, correct for ordinal
# 1-5 Likert scores on paired samples). Reports p-values and Cohen's d effect size.

import numpy as np
from scipy.stats import wilcoxon

def _extract_paired_cmqs(scores_a: list, scores_b: list):
    """Return parallel CMQS arrays for records matched by chunk_id+slot_index."""
    map_a = {f"{r.get('chunk_id')}_{r.get('slot_index')}": r.get("CMQS")
             for r in scores_a if r.get("status") == "ok" and r.get("CMQS") is not None}
    map_b = {f"{r.get('chunk_id')}_{r.get('slot_index')}": r.get("CMQS")
             for r in scores_b if r.get("status") == "ok" and r.get("CMQS") is not None}
    common = sorted(map_a.keys() & map_b.keys())
    return (np.array([map_a[k] for k in common]),
            np.array([map_b[k] for k in common]),
            len(common))

def _extract_paired_dim(scores_a: list, scores_b: list, dim_key: str):
    """Return parallel arrays for a single dimension."""
    def _score(r):
        d = (r.get("scores") or {}).get(dim_key, {})
        return d.get("score") if isinstance(d, dict) else d
    map_a = {f"{r.get('chunk_id')}_{r.get('slot_index')}": _score(r)
             for r in scores_a if r.get("status") == "ok" and _score(r) is not None}
    map_b = {f"{r.get('chunk_id')}_{r.get('slot_index')}": _score(r)
             for r in scores_b if r.get("status") == "ok" and _score(r) is not None}
    common = sorted(map_a.keys() & map_b.keys())
    return (np.array([float(map_a[k]) for k in common]),
            np.array([float(map_b[k]) for k in common]),
            len(common))

def cohens_d(a: np.ndarray, b: np.ndarray) -> float:
    """Cohen's d = (mean_a - mean_b) / pooled_std."""
    pooled = np.sqrt((np.std(a, ddof=1)**2 + np.std(b, ddof=1)**2) / 2)
    return float((np.mean(a) - np.mean(b)) / pooled) if pooled > 0 else 0.0

def interpret_d(d: float) -> str:
    ad = abs(d)
    if ad < 0.2:  return "negligible"
    if ad < 0.5:  return "small"
    if ad < 0.8:  return "medium"
    return "large"

def wilcoxon_test(a: np.ndarray, b: np.ndarray, alternative: str = "greater"):
    """Run Wilcoxon signed-rank. Returns (stat, p, n)."""
    if len(a) < 10:
        return None, None, len(a)
    try:
        stat, p = wilcoxon(a, b, alternative=alternative, zero_method="wilcox")
        return float(stat), float(p), len(a)
    except Exception as e:
        return None, None, len(a)

# ── Run tests ─────────────────────────────────────────────────────────────────
print("\n" + "="*72)
print("STATISTICAL SIGNIFICANCE TESTS — Wilcoxon Signed-Rank + Cohen's d")
print("="*72)
print(f"  H0: no difference between paired samples (one-sided, alternative='greater')")
print(f"  α  = 0.05  |  Effect size: negligible <0.2, small 0.2, medium 0.5, large 0.8")
print()

_significance_results = {}

# CMQS: Pre-FT vs Post-FT
a_cmqs, b_cmqs, n1 = _extract_paired_cmqs(student_post_scores, student_pre_scores)
stat, p, n = wilcoxon_test(a_cmqs, b_cmqs)
d = cohens_d(a_cmqs, b_cmqs)
_significance_results["cmqs_post_vs_pre"] = {"stat": stat, "p": p, "n": n, "cohens_d": round(d,4)}
p_str = f"{p:.4f}" if p is not None else "N/A"
sig   = "✓ significant" if (p is not None and p < 0.05) else "✗ not significant"
print(f"  CMQS  Post-FT > Pre-FT      : W={stat}, p={p_str}  d={d:+.3f} ({interpret_d(d)})  [{sig}]  n={n}")

# CMQS: Teacher vs Post-FT (Post-FT < Teacher, so we test alternative='less')
a_cmqs_t, b_cmqs_p, n2 = _extract_paired_cmqs(teacher_scores, student_post_scores)
stat2, p2, n2 = wilcoxon_test(a_cmqs_t, b_cmqs_p, alternative="greater")
d2 = cohens_d(a_cmqs_t, b_cmqs_p)
_significance_results["cmqs_teacher_vs_post"] = {"stat": stat2, "p": p2, "n": n2, "cohens_d": round(d2,4)}
p2_str = f"{p2:.4f}" if p2 is not None else "N/A"
sig2   = "✓ significant" if (p2 is not None and p2 < 0.05) else "✗ not significant"
print(f"  CMQS  Teacher > Post-FT     : W={stat2}, p={p2_str}  d={d2:+.3f} ({interpret_d(d2)})  [{sig2}]  n={n2}")

print()
print(f"  {'Dimension':<36} {'Post>Pre W':>10}  {'p-val':>7}  {'d':>6}  {'Effect':>10}  {'Sig?':>15}")
print("  " + "─"*85)

for key in DIM_KEYS:
    lbl = DIM_LABELS[key]
    a_d, b_d, nd = _extract_paired_dim(student_post_scores, student_pre_scores, key)
    stat_d, p_d, nd = wilcoxon_test(a_d, b_d)
    cd = cohens_d(a_d, b_d)
    _significance_results[key] = {"stat": stat_d, "p": p_d, "n": nd, "cohens_d": round(cd, 4)}
    pd_str = f"{p_d:.4f}" if p_d is not None else "  N/A "
    sig_d  = "✓ sig" if (p_d is not None and p_d < 0.05) else "✗ n.s."
    print(f"  {lbl:<36} {str(stat_d):>10}  {pd_str:>7}  {cd:>+6.3f}  {interpret_d(cd):>10}  {sig_d:>15}  n={nd}")

print()
print("  Note: Wilcoxon signed-rank on matched pairs (same chunk_id+slot_index across stages).")
print("        Post-FT > Pre-FT tests confirm fine-tuning improved quality.")
print("        Teacher > Post-FT tests confirm the residual quality gap.")

# Save
_stats_extended = all_stats.copy()
_stats_extended["significance_tests"] = _significance_results
with open(COMPARISON_PATH, "w", encoding="utf-8") as f:
    json.dump(_stats_extended, f, indent=2, ensure_ascii=False)
print(f"\n✓ Significance results appended to → {COMPARISON_PATH}")
````

### 12.13  P3 — Sentence-BERT Teacher-Student Semantic Similarity

Computes **pairwise cosine similarity** between teacher and student MCQ embeddings using `bge-large-en-v1.5`.

This is a **judge-free** signal — it does not rely on the LLM judge at all. High similarity means the student generates questions that are semantically close to the teacher's, independent of any scoring rubric.

| Pair | Interpretation |
|---|---|
| Teacher ↔ Pre-FT | Baseline semantic distance before distillation |
| Teacher ↔ Post-FT | Semantic distance after distillation |

A positive similarity gain (Post > Pre) confirms that fine-tuning moved the student's generation distribution closer to the teacher's.

Outputs:
- Summary table (mean, std, n per pair)
- Bloom-level breakdown of Post-FT similarity
- **V8** `semantic_similarity.png` — histogram distributions + Bloom line chart


````python
# ─── 12.13  P3 — Sentence-BERT Teacher-Student Semantic Similarity ──────────

try:
    from sentence_transformers import SentenceTransformer
    from sklearn.metrics.pairwise import cosine_similarity as cos_sim
    _sbert_available = True
except ImportError:
    import subprocess, sys
    subprocess.check_call([
        sys.executable, "-m", "pip", "install", "-q",
        "sentence-transformers", "scikit-learn"
    ])
    from sentence_transformers import SentenceTransformer
    from sklearn.metrics.pairwise import cosine_similarity as cos_sim
    _sbert_available = True

import numpy as np
import random
import logging
import matplotlib.pyplot as plt

logging.getLogger("sentence_transformers").setLevel(logging.ERROR)
logging.getLogger("transformers").setLevel(logging.ERROR)
logging.getLogger("transformers.modeling_utils").setLevel(logging.ERROR)

# ── Build text lookup from OUTPUT JSONL files ────────────────────────────────
# Score records contain only scores/CMQS.
# Actual MCQ text lives in the output files.

def _build_text_lookup(output_path: str) -> dict:
    """
    Build:
        {chunk_id_slotindex: 'question optA optB ...'}
    """
    lookup = {}

    for r in load_jsonl_safe(output_path):

        mcq  = r.get("mcq") or {}
        q    = mcq.get("question", "").strip()
        opts = mcq.get("options") or {}

        if not q:
            continue

        opts_text = " ".join(
            f"{k}: {v}"
            for k, v in opts.items()
        )

        key = f"{r.get('chunk_id', '')}_{r.get('slot_index', '')}"

        lookup[key] = f"{q} {opts_text}".strip()

    return lookup


print("Building MCQ text lookups from output files...")

_teacher_text_lookup = _build_text_lookup(TEACHER_OUTPUT_PATH)
_pre_ft_text_lookup  = _build_text_lookup(STUDENT_PRE_OUTPUT_PATH)
_post_ft_text_lookup = _build_text_lookup(STUDENT_POST_OUTPUT_PATH)

print(f"  teacher_outputs  : {len(_teacher_text_lookup):,} entries")
print(f"  pre_ft_outputs   : {len(_pre_ft_text_lookup):,} entries")
print(f"  post_ft_outputs  : {len(_post_ft_text_lookup):,} entries")


# ── Build valid MCQ maps (status=ok only) ────────────────────────────────────

def _build_mcq_map(score_records: list, text_lookup: dict) -> dict:

    out = {}

    for r in score_records:

        if r.get("status") != "ok":
            continue

        k = f"{r.get('chunk_id')}_{r.get('slot_index')}"

        text = text_lookup.get(k, "").strip()

        if text:
            out[k] = text

    return out


print("\nLoading Sentence-BERT model (bge-large-en-v1.5)...")

_sbert = SentenceTransformer("BAAI/bge-large-en-v1.5")

map_teacher = _build_mcq_map(teacher_scores,      _teacher_text_lookup)
map_pre_ft  = _build_mcq_map(student_pre_scores,  _pre_ft_text_lookup)
map_post_ft = _build_mcq_map(student_post_scores, _post_ft_text_lookup)

print(f"  map_teacher : {len(map_teacher):,} entries")
print(f"  map_pre_ft  : {len(map_pre_ft):,} entries")
print(f"  map_post_ft : {len(map_post_ft):,} entries")


# ══════════════════════════════════════════════════════════════════════════════
# Match ONLY shared samples across Teacher / Pre-FT / Post-FT
# This ensures a fair apples-to-apples comparison.
# ══════════════════════════════════════════════════════════════════════════════

common_all = sorted(
    map_teacher.keys()
    & map_pre_ft.keys()
    & map_post_ft.keys()
)

print(f"\nShared matched pairs across all stages: {len(common_all)}")

MAX_MATCHED_SAMPLES = 150

if len(common_all) > MAX_MATCHED_SAMPLES:

    rng = random.Random(42)

    selected_keys = sorted(
        rng.sample(common_all, MAX_MATCHED_SAMPLES)
    )

else:
    selected_keys = common_all

print(f"Using matched shared pairs: {len(selected_keys)}")


# ── Build aligned text lists ─────────────────────────────────────────────────

teacher_texts = [map_teacher[k] for k in selected_keys]
pre_ft_texts  = [map_pre_ft[k]  for k in selected_keys]
post_ft_texts = [map_post_ft[k] for k in selected_keys]


# ── Encode embeddings ────────────────────────────────────────────────────────

teacher_embs = _sbert.encode(
    teacher_texts,
    batch_size=64,
    show_progress_bar=False
)

pre_ft_embs = _sbert.encode(
    pre_ft_texts,
    batch_size=64,
    show_progress_bar=False
)

post_ft_embs = _sbert.encode(
    post_ft_texts,
    batch_size=64,
    show_progress_bar=False
)


# ── Pairwise cosine similarities ─────────────────────────────────────────────

pre_sims = [
    float(cos_sim([teacher_embs[i]], [pre_ft_embs[i]])[0][0])
    for i in range(len(selected_keys))
]

post_sims = [
    float(cos_sim([teacher_embs[i]], [post_ft_embs[i]])[0][0])
    for i in range(len(selected_keys))
]


# ── Aggregate helper ─────────────────────────────────────────────────────────

def _agg(vals):

    if not vals:
        return {}

    a = np.array(vals)

    return {
        "mean":   round(float(np.mean(a)),   4),
        "std":    round(float(np.std(a)),    4),
        "median": round(float(np.median(a)), 4),
        "min":    round(float(np.min(a)),    4),
        "max":    round(float(np.max(a)),    4),
        "n":      len(a),
    }


pre_agg  = _agg(pre_sims)
post_agg = _agg(post_sims)


# ── Main results ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("SENTENCE-BERT COSINE SIMILARITY (Teacher ↔ Student)")
print("=" * 60)

print("  Model: bge-large-en-v1.5 | Metric: pairwise cosine similarity")
print()

print(f"  {'Stage':<28} {'Mean':>7}  {'Std':>6}  {'Median':>8}  {'n':>5}")
print("  " + "─" * 55)

print(
    f"  {'Teacher ↔ Pre-FT Student':<28} "
    f"{pre_agg['mean']:>7.4f}  "
    f"{pre_agg['std']:>6.4f}  "
    f"{pre_agg['median']:>8.4f}  "
    f"{pre_agg['n']:>5}"
)

print(
    f"  {'Teacher ↔ Post-FT Student':<28} "
    f"{post_agg['mean']:>7.4f}  "
    f"{post_agg['std']:>6.4f}  "
    f"{post_agg['median']:>8.4f}  "
    f"{post_agg['n']:>5}"
)

delta = round(post_agg["mean"] - pre_agg["mean"], 4)

print()
print(f"  Similarity gain (Post-FT − Pre-FT): {delta:+.4f}")

if delta > 0:
    print("  Interpretation: Fine-tuning improved semantic alignment with teacher.")
else:
    print("  Interpretation: No improvement in semantic alignment.")


# ── Bloom-level breakdown ────────────────────────────────────────────────────

print("\n  Post-FT similarity by Bloom level:")

_key_bloom = {}

for r in student_post_scores:

    k = f"{r.get('chunk_id')}_{r.get('slot_index')}"

    _key_bloom[k] = r.get("bloom_level", "unknown")


bloom_sim = {}

for k, sim in zip(selected_keys, post_sims):

    bl = _key_bloom.get(k, "unknown")

    bloom_sim.setdefault(bl, []).append(sim)


for bl in BLOOM_LEVELS:

    vals = bloom_sim.get(bl, [])

    if vals:
        print(f"    {bl:<12}: {np.mean(vals):.4f}  (n={len(vals)})")


# ── Visualization ────────────────────────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(
    pre_sims,
    bins=20,
    alpha=0.65,
    color="#DC2626",
    label="Teacher ↔ Pre-FT"
)

axes[0].hist(
    post_sims,
    bins=20,
    alpha=0.65,
    color="#16A34A",
    label="Teacher ↔ Post-FT"
)

axes[0].set_xlabel("Cosine Similarity", fontsize=11)
axes[0].set_ylabel("Count", fontsize=11)

axes[0].set_title(
    "Semantic Similarity Distributions\n(Teacher ↔ Student)",
    fontsize=11
)

axes[0].legend(fontsize=9)

axes[0].axvline(
    pre_agg["mean"],
    color="#DC2626",
    ls="--",
    lw=1.5
)

axes[0].axvline(
    post_agg["mean"],
    color="#16A34A",
    ls="--",
    lw=1.5
)

axes[0].grid(alpha=0.3)


bl_means = [
    np.mean(bloom_sim.get(bl, [np.nan]))
    for bl in BLOOM_LEVELS
]

axes[1].plot(
    BLOOM_LEVELS,
    bl_means,
    "o-",
    color="#16A34A",
    linewidth=2.5,
    markersize=7
)

axes[1].set_ylim(0, 1)

axes[1].set_ylabel("Mean Cosine Similarity", fontsize=11)
axes[1].set_xlabel("Bloom Level", fontsize=11)

axes[1].set_title(
    "Post-FT Semantic Similarity\nby Bloom Level",
    fontsize=11
)

axes[1].grid(alpha=0.3)

plt.tight_layout()

_p3_path = os.path.join(PLOT_DIR, "V8_semantic_similarity.png")

plt.savefig(_p3_path, dpi=150)

plt.close()

print(f"\n✓ Semantic similarity chart saved → {_p3_path}")


# ── Save ─────────────────────────────────────────────────────────────────────

_stats_extended["semantic_similarity"] = {
    "matched_shared_pairs": len(selected_keys),

    "pre_ft": pre_agg,

    "post_ft": post_agg,

    "similarity_gain": delta,

    "bloom_breakdown_post_ft": {
        bl: round(float(np.mean(v)), 4)
        for bl, v in bloom_sim.items()
    },
}

with open(COMPARISON_PATH, "w", encoding="utf-8") as f:

    json.dump(_stats_extended, f, indent=2, ensure_ascii=False)

print(f"✓ Semantic similarity results saved → {COMPARISON_PATH}")
````

### 12.14  P4 — Dimension Gap Ranking & Failure Mode Analysis

Ranks all 6 evaluation dimensions by the **Teacher − Post-FT gap** (largest gap first) to identify **knowledge-transfer bottlenecks** — dimensions where the student failed to absorb the teacher's quality.

Gap severity thresholds:

| Gap | Severity |
|---|---|
| < 0.25 | ✓ Negligible — excellent transfer |
| 0.25 – 0.50 | ~ Small — acceptable |
| 0.50 – 0.75 | ! Moderate — needs attention |
| ≥ 0.75 | ✗ Large — transfer failure |

A special note is printed if D4 (Distractor Plausibility) has a small gap but a low absolute teacher score — indicating a system-wide ceiling rather than a student failure.

Outputs:
- Ranked table with gap, FT Gain, Post-QPS, and severity per dimension
- **V9** `gap_ranking.png` — horizontal bar chart coloured by severity


````python
# ─── 12.14  P4 — Dimension Gap Ranking & Failure Mode Analysis ───────────────
# Ranks all 6 dimensions by Teacher–Post-FT gap size.
# Identifies which dimension is the biggest knowledge-transfer bottleneck.

import matplotlib.pyplot as plt
import numpy as np

all_keys_p4  = DIM_KEYS + ["CMQS"]
all_labels_p4 = {**DIM_LABELS, "CMQS": "CMQS (Composite)"}

gaps = []
for key in all_keys_p4:
    t_mean = stats_teacher.get(key, {}).get("mean", 0)
    a_mean = stats_post_ft.get(key, {}).get("mean", 0)
    b_mean = stats_pre_ft.get(key,  {}).get("mean", 0)
    gap    = round(t_mean - a_mean, 4)
    gain   = round(a_mean - b_mean, 4)
    post_q = round(a_mean / t_mean * 100, 1) if t_mean else 0
    gaps.append({
        "key":     key,
        "label":   all_labels_p4[key],
        "teacher": t_mean,
        "pre_ft":  b_mean,
        "post_ft": a_mean,
        "gap":     gap,
        "gain":    gain,
        "post_qps": post_q,
    })

gaps_ranked = sorted(gaps, key=lambda x: x["gap"], reverse=True)

print("\n" + "="*75)
print("DIMENSION GAP RANKING — Teacher vs Post-FT Student (Largest Gap First)")
print("="*75)
print(f"  {'Dimension':<36} {'Teacher':>8} {'Post-FT':>8} {'Gap (↓bad)':>11} {'FT Gain':>9} {'Post-QPS':>9}")
print("  " + "─"*75)

for row in gaps_ranked:
    if row["gap"] < 0.25:   severity = "✓ negligible"
    elif row["gap"] < 0.50: severity = "~ small"
    elif row["gap"] < 0.75: severity = "! moderate"
    else:                   severity = "✗ LARGE"
    print(f"  {row['label']:<36} {row['teacher']:>8.3f} {row['post_ft']:>8.3f} "
          f"{row['gap']:>+11.3f} {row['gain']:>+9.3f} {str(row['post_qps'])+'%':>9}  {severity}")

print()
print("  Gap interpretation: <0.25 negligible | 0.25-0.50 small | 0.50-0.75 moderate | >0.75 large")

# ── Special D4 floor note ─────────────────────────────────────────────────────
d4_row = next((r for r in gaps if r["key"] == "D4_distractor_plausibility"), None)
if d4_row and d4_row["teacher"] < 4.2:
    print(f"\n  NOTE — D4 (Distractor Plausibility):")
    print(f"    Teacher absolute score = {d4_row['teacher']:.3f} — the teacher itself does not score high here.")
    print(f"    Gap = {d4_row['gap']:.3f} (small), meaning this is a system-wide ceiling, not a transfer failure.")
    print(f"    Distractor plausibility is the hardest MCQ generation sub-task; the small gap confirms")
    print(f"    the student successfully distilled this capability to the teacher's own level.")

# ── Horizontal bar chart ──────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 5))
labels_sorted = [r["label"] for r in gaps_ranked]
gap_vals      = [r["gap"]   for r in gaps_ranked]
colors_gap    = ["#DC2626" if g > 0.5 else "#F59E0B" if g > 0.25 else "#16A34A"
                 for g in gap_vals]

bars = ax.barh(labels_sorted, gap_vals, color=colors_gap, alpha=0.85)
for bar, val in zip(bars, gap_vals):
    ax.text(val + 0.005, bar.get_y() + bar.get_height()/2,
            f"{val:+.3f}", va="center", fontsize=10)

ax.set_xlabel("Teacher − Post-FT Gap (lower = better transfer)", fontsize=11)
ax.set_title("Dimension Gap Ranking — Knowledge Transfer Failure Modes", fontsize=12)
ax.axvline(0.25, color="goldenrod", ls="--", lw=1.2, alpha=0.7, label="Small gap threshold (0.25)")
ax.axvline(0.50, color="#DC2626",   ls="--", lw=1.2, alpha=0.7, label="Moderate gap threshold (0.50)")
ax.legend(fontsize=9)
ax.grid(axis="x", alpha=0.3)
ax.invert_yaxis()
plt.tight_layout()
_p4_path = os.path.join(PLOT_DIR, "V9_gap_ranking.png")
plt.savefig(_p4_path, dpi=150, bbox_inches="tight")
plt.close()
print(f"\n✓ Gap ranking chart saved → {_p4_path}")

_stats_extended["gap_ranking"] = gaps_ranked
with open(COMPARISON_PATH, "w", encoding="utf-8") as f:
    json.dump(_stats_extended, f, indent=2, ensure_ascii=False)
````

### 12.15  P5 — Retention Score & Efficiency Analysis

Computes the **Retention Score (RS)** — a single number capturing the quality-per-compute trade-off:

```
RS = (Post-FT CMQS / Teacher CMQS) × (Teacher Latency / Student Latency)
```

RS > 1.0 means the student delivers **better quality-per-compute** than the teacher, justifying the knowledge distillation approach.

| Component | Meaning |
|---|---|
| CMQS ratio | How much quality the student retains |
| Latency ratio | How much faster the student is |
| Model compression | How many fewer parameters the student has |

> Student latency is measured live if the post-FT model is in scope; otherwise representative estimates are used (Qwen-1.5B on T4 GPU ≈ 320 ms, Gemini API ≈ 1800 ms round-trip).

Outputs:
- Summary table (params, latency, CMQS, RS, speedup, compression)
- **V10** `retention_score.png` — side-by-side bars for params / latency / CMQS


````python
# ─── 12.15  P5 — Retention Score & Efficiency Analysis ───────────────────────
# RS = (Student_Post_CMQS / Teacher_CMQS) × (Teacher_Latency / Student_Latency)
# RS > 1.0 means the student achieves better quality-per-compute than the teacher.
#
# Latency is measured on a REAL sample from the held-out test set using the
# exact SFT prompt template (SFT_SYSTEM_PROMPT + build_user_prompt_conditioned).
# Student  → local HF inference (Rehab-Hamdy/mcq-bloom-qwen-merged_v4).
# Teacher  → live Gemini 2.5 Flash Lite API call on the same prompt.
# ─────────────────────────────────────────────────────────────────────────────

import time, gc, os
import torch
import numpy as np
import matplotlib.pyplot as plt
from typing import Optional

# ── Constants ─────────────────────────────────────────────────────────────────
POST_FT_HF_REPO   = "Rehab-Hamdy/mcq-bloom-qwen-merged_v4"
GEMINI_MODEL_NAME = "gemini-2.5-flash-lite"
_N_WARMUP         = 2
_N_MEASURE        = 5     # repetitions for a stable mean latency

# ══════════════════════════════════════════════════════════════════════════════
# STEP 1 — Select one real sample from the held-out test set
# ══════════════════════════════════════════════════════════════════════════════
_latency_test_rec  = None
_latency_test_slot = None

import random as _random
_rng = _random.Random(42)
_candidates = list(teacher_eval_records)
_rng.shuffle(_candidates)

for _rec in _candidates:
    _cid     = _rec.get("chunk_id", "")
    _chunk   = chunk_lookup.get(_cid, {})
    if not _chunk:
        continue
    _s3_meta = _chunk.get("metadata", {})
    _fringe  = _chunk.get("context_fringe", {})
    _slot = {
        "text":          _chunk.get("text", _rec.get("text", "")),
        "bloom_level":   _rec.get("bloom_level",  _s3_meta.get("bloom_level",  "understand")),
        "chunk_type":    _rec.get("chunk_type",   _s3_meta.get("chunk_type",   "definition")),
        "concept":       _rec.get("concept",      (_s3_meta.get("concepts") or [""])[0]),
        "keywords":      _s3_meta.get("keywords", []),
        "prev_sentence": (_fringe.get("prev_sentence") or "").strip(),
        "next_sentence": (_fringe.get("next_sentence") or "").strip(),
    }
    if _slot["text"].strip():
        _latency_test_rec  = _rec
        _latency_test_slot = _slot
        break

if _latency_test_slot is None:
    raise RuntimeError(
        "Could not find a usable sample from teacher_eval_records. "
        "Ensure chunk_lookup is populated (cell 12.5 must have run)."
    )

# Build the real prompt using the pipeline's exact template
_real_user_prompt = build_user_prompt_conditioned(_latency_test_slot)
_real_messages    = [
    {"role": "system", "content": SFT_SYSTEM_PROMPT},
    {"role": "user",   "content": _real_user_prompt},
]

# Dynamic sample description reused in all prints, chart, and JSON
_sample_desc = (
    f"chunk_id={_latency_test_rec.get('chunk_id', '?')} | "
    f"slot={_latency_test_rec.get('slot_index', '?')} | "
    f"bloom={_latency_test_slot['bloom_level']} | "
    f"chunk_type={_latency_test_slot['chunk_type']} | "
    f"concept='{_latency_test_slot['concept']}'"
)

print("=" * 70)
print("LATENCY TEST SAMPLE  (real record from held-out test set)")
print("=" * 70)
print(f"  {_sample_desc}")
print(f"  chunk chars : {len(_latency_test_slot['text'])}")
print(f"  prompt chars: {len(_real_user_prompt)}")
print()
print("  NOTE: All latency figures below are measured on this real test sample")
print("        using the exact SFT prompt template as the quiz generation pipeline.")
print("=" * 70)

# ══════════════════════════════════════════════════════════════════════════════
# STEP 2 — Student model latency (local HF inference)
# ══════════════════════════════════════════════════════════════════════════════

def _measure_student_latency_ms(model, tokenizer, messages, n_warmup=2, n_measure=5):
    """Measure mean generation latency using apply_chat_template (same as
    _run_student_inference in 12.6/12.7) on the real prompt messages."""
    text   = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer([text], return_tensors="pt").to(model.device)
    for _ in range(n_warmup):
        with torch.no_grad():
            model.generate(**inputs, max_new_tokens=256, do_sample=False,
                           temperature=None, top_p=None)
    if torch.cuda.is_available():
        torch.cuda.synchronize()
    times = []
    for _ in range(n_measure):
        t0 = time.perf_counter()
        with torch.no_grad():
            model.generate(**inputs, max_new_tokens=256, do_sample=False,
                           temperature=None, top_p=None)
        if torch.cuda.is_available():
            torch.cuda.synchronize()
        times.append((time.perf_counter() - t0) * 1000)
    return float(np.mean(times))


_student_latency_ms   = None
_student_model_source = "unknown"

# ── 2a: Use model already in kernel scope (from 12.7 / 12.6) ─────────────────
for _mvar, _tvar, _label in [
    ("post_model", "post_tokenizer", "post_model (12.7)"),
    ("pre_model",  "pre_tokenizer",  "pre_model  (12.6)"),
]:
    if _mvar in dir() and _tvar in dir():
        try:
            print(f"Measuring student latency using {_label}...")
            _student_latency_ms = _measure_student_latency_ms(
                eval(_mvar), eval(_tvar), _real_messages, _N_WARMUP, _N_MEASURE
            )
            _student_model_source = _label
            print(f"  Student latency : {_student_latency_ms:.1f} ms  "
                  f"(mean over {_N_MEASURE} runs)")
            break
        except Exception as _e:
            print(f"  Could not measure {_label}: {_e}")

# ── 2b: Load post-FT model from HF if not in scope ────────────────────────────
if _student_latency_ms is None:
    print(f"\nNo model in scope — loading post-FT student from HF: {POST_FT_HF_REPO}")
    try:
        from transformers import AutoModelForCausalLM, AutoTokenizer
        DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
        _lat_tokenizer = AutoTokenizer.from_pretrained(
            POST_FT_HF_REPO, token=HF_TOKEN, trust_remote_code=True
        )
        _lat_model = AutoModelForCausalLM.from_pretrained(
            POST_FT_HF_REPO,
            torch_dtype=torch.bfloat16,
            device_map="auto",
            token=HF_TOKEN,
            trust_remote_code=True,
        )
        _lat_model.eval()
        print(f"  Model loaded on {DEVICE}")

        _student_latency_ms = _measure_student_latency_ms(
            _lat_model, _lat_tokenizer, _real_messages, _N_WARMUP, _N_MEASURE
        )
        _student_model_source = POST_FT_HF_REPO
        print(f"  Student latency : {_student_latency_ms:.1f} ms  "
              f"(mean over {_N_MEASURE} runs)")

        del _lat_model
        gc.collect()
        torch.cuda.empty_cache()
        print("  Post-FT model unloaded from GPU.")
    except Exception as _e:
        print(f"  Failed to load model from HF: {_e}")
        print("  Falling back to representative estimate (320 ms).")

# ── 2c: Last-resort fallback ──────────────────────────────────────────────────
if _student_latency_ms is None:
    _student_latency_ms   = 320.0
    _student_model_source = "estimated (Qwen-1.5B on T4 GPU — model load failed)"

# ══════════════════════════════════════════════════════════════════════════════
# STEP 3 — Teacher (Gemini 2.5 Flash Lite) latency via live API call
# ══════════════════════════════════════════════════════════════════════════════

def _measure_gemini_latency_ms(prompt_messages, model_name=GEMINI_MODEL_NAME, n_measure=5):
    """Measure mean wall-clock API latency for Gemini 2.5 Flash Lite.
    Each call sends the real prompt and waits for the full response."""
    try:
        import google.generativeai as genai
    except ImportError:
        import subprocess, sys
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q",
                               "google-generativeai"])
        import google.generativeai as genai

    # Always read the latest key from Kaggle Secrets
    try:
        from kaggle_secrets import UserSecretsClient
        api_key = UserSecretsClient().get_secret("GEMINI_API_KEY")
    except Exception:
        api_key = ""
    
    if not api_key:
        raise ValueError(
            "GEMINI_API_KEY not found. Add it as a Kaggle Secret named "
            "'GEMINI_API_KEY' (Settings -> Add-ons -> Secrets)."
        )

    genai.configure(api_key=api_key)
    _gmodel = genai.GenerativeModel(
        model_name=model_name,
        system_instruction=prompt_messages[0]["content"],
    )
    _user_text = prompt_messages[1]["content"]

    times = []
    for i in range(n_measure):
        t0 = time.perf_counter()
        _gmodel.generate_content(
            _user_text,
            generation_config=genai.GenerationConfig(
                temperature=0.0,
                max_output_tokens=512,
            ),
        )
        times.append((time.perf_counter() - t0) * 1000)
        if i < n_measure - 1:
            time.sleep(0.5)
    return float(np.mean(times))


_teacher_latency_ms = None
_teacher_source     = "unknown"

print(f"\nMeasuring teacher (Gemini) latency via API ({GEMINI_MODEL_NAME})...")
try:
    _teacher_latency_ms = _measure_gemini_latency_ms(
        _real_messages, model_name=GEMINI_MODEL_NAME, n_measure=_N_MEASURE
    )
    _teacher_source = f"measured — {GEMINI_MODEL_NAME} API"
    print(f"  Teacher latency : {_teacher_latency_ms:.1f} ms  "
          f"(mean over {_N_MEASURE} runs)")
except Exception as _e:
    print(f"  Gemini API call failed: {_e}")
    print("  Falling back to representative estimate (1800 ms).")
    _teacher_latency_ms = 1800.0
    _teacher_source = f"estimated ({GEMINI_MODEL_NAME} API unavailable)"

_latency_source = (
    f"real test sample | sample: [{_sample_desc}] | "
    f"student: {_student_model_source} | teacher: {_teacher_source}"
)

# ══════════════════════════════════════════════════════════════════════════════
# STEP 4 — Compute Retention Score
# ══════════════════════════════════════════════════════════════════════════════
_cmqs_ratio    = post_ft_cmqs / teacher_cmqs if teacher_cmqs else 0.0
# latency_ratio > 1 → student faster; < 1 → student slower
_latency_ratio = _teacher_latency_ms / _student_latency_ms if _student_latency_ms else 0.0
_RS            = _cmqs_ratio * _latency_ratio

# ── Human-readable speed direction ───────
if _latency_ratio >= 1.0:
    _speed_label = f"{_latency_ratio:.2f}x Student FASTER than teacher"
else:
    # student is slower: report how many times slower it is
    _slower_ratio = _student_latency_ms / _teacher_latency_ms
    _speed_label  = f"{_slower_ratio:.2f}x Student SLOWER than teacher"

print("\n" + "=" * 70)
print("RETENTION SCORE & EFFICIENCY ANALYSIS")
print("=" * 70)
print(f"  Sample             : {_sample_desc}")
print(f"  Student source     : {_student_model_source}")
print(f"  Teacher source     : {_teacher_source}")
print()
print(f"  {'Metric':<40} {'Teacher':>14} {'Post-FT Student':>16}")
print("  " + "\u2500" * 72)
print(f"  {'Inference Latency (ms)':<40} {_teacher_latency_ms:>14.1f} {_student_latency_ms:>16.1f}")
print(f"  {'CMQS (mean)':<40} {teacher_cmqs:>14.4f} {post_ft_cmqs:>16.4f}")
print()
print(f"  Latency comparison : {_speed_label}")
print(f"  CMQS Retention     : {_cmqs_ratio*100:.1f}%  "
      f"(student retains {_cmqs_ratio*100:.1f}% of teacher quality)")
print(f"  Latency ratio (T/S): {_latency_ratio:.2f}  "
      f"(>1 = student faster, <1 = student slower)")
print(f"  Retention Score (RS) : {_RS:.4f}")
print()
if _RS >= 1.0:
    print(f"  RS = {_RS:.4f} >= 1.0  Student achieves BETTER quality-per-compute than the teacher.")
    print(f"  Knowledge distillation efficiency gain: {_RS:.2f}x overall.")
else:
    # Distinguish whether the RS < 1 is due to quality loss, speed, or both
    _quality_ok = _cmqs_ratio >= 1.0
    _speed_ok   = _latency_ratio >= 1.0
    if not _speed_ok and _quality_ok:
        print(f"  RS = {_RS:.4f} < 1.0  Quality is preserved but the student is slower than the teacher.")
        print(f"  The student is {_slower_ratio:.2f}x slower, which pulls RS below 1.0.")
        print(f"  Consider optimising inference (quantisation, batching) to recover efficiency.")
    else:
        print(f"  RS = {_RS:.4f} < 1.0  Quality loss outweighs the efficiency gain.")

print()
print(f"  NOTE: Latency measured on a real held-out test sample.")
print(f"        Sample      : {_sample_desc}")
print(f"        Repetitions : {_N_WARMUP} warmup + {_N_MEASURE} timed runs per model.")

# ══════════════════════════════════════════════════════════════════════════════
# STEP 5 — Bar chart (V10) — Latency & CMQS
# ══════════════════════════════════════════════════════════════════════════════
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

_teacher_label = f"Teacher\n({GEMINI_MODEL_NAME})"
_student_label = f"Post-FT Student\n({POST_FT_HF_REPO.split('/')[-1]})"

for ax, (title, vals, labels, colors) in zip(axes, [
    ("Inference Latency (ms)",
     [_teacher_latency_ms, _student_latency_ms],
     [_teacher_label, _student_label],
     ["#2563EB", "#16A34A"]),
    ("CMQS Score",
     [teacher_cmqs, post_ft_cmqs],
     [_teacher_label, _student_label],
     ["#2563EB", "#16A34A"]),
]):
    bars = ax.bar(labels, vals, color=colors, alpha=0.85, width=0.45)
    for bar, v in zip(bars, vals):
        ax.text(bar.get_x() + bar.get_width() / 2,
                v + max(vals) * 0.02,
                f"{v:.1f}" if v > 10 else f"{v:.4f}",
                ha="center", va="bottom", fontsize=11)
    ax.set_title(title, fontsize=11)
    ax.grid(axis="y", alpha=0.3)
    ax.set_ylim(0, max(vals) * 1.2)

plt.suptitle(
    f"Efficiency-Quality Trade-off  |  RS = {_RS:.4f}  |  {_speed_label}\n"
    f"Latency measured on real test sample: {_sample_desc}",
    fontsize=9, y=1.04,
)
plt.tight_layout()
_p6_path = os.path.join(PLOT_DIR, "V10_retention_score.png")
plt.savefig(_p6_path, dpi=150, bbox_inches="tight")
plt.close()
print(f"\n  Retention score chart saved -> {_p6_path}")

# ══════════════════════════════════════════════════════════════════════════════
# STEP 6 — Save to final_comparison_results.json
# ══════════════════════════════════════════════════════════════════════════════
_stats_extended["retention_score"] = {
    "teacher_latency_ms":    round(_teacher_latency_ms, 2),
    "student_latency_ms":    round(_student_latency_ms, 2),
    "latency_ratio_T_over_S":round(_latency_ratio, 4),
    "speed_description":     _speed_label,
    "cmqs_retention_pct":    round(_cmqs_ratio * 100, 2),
    "retention_score_RS":    round(_RS, 4),
    "latency_source":        _latency_source,
    "latency_sample": {
        "chunk_id":    _latency_test_rec.get("chunk_id", ""),
        "slot_index":  _latency_test_rec.get("slot_index", 0),
        "bloom_level": _latency_test_slot["bloom_level"],
        "chunk_type":  _latency_test_slot["chunk_type"],
        "concept":     _latency_test_slot["concept"],
        "prompt_chars":len(_real_user_prompt),
        "n_warmup":    _N_WARMUP,
        "n_measure":   _N_MEASURE,
        "student_model": _student_model_source,
        "teacher_model": _teacher_source,
    },
}

with open(COMPARISON_PATH, "w", encoding="utf-8") as f:
    json.dump(_stats_extended, f, indent=2, ensure_ascii=False)
print(f"  Retention score results saved -> {COMPARISON_PATH}")
````

### 12.16  P6 — KL Divergence on Score Distributions (All 6 Dimensions)

Extends beyond mean comparisons by measuring how the **full score distribution** of the student differs from the teacher's, per dimension.

**Metric:** KL divergence KL(P_teacher ∥ Q_student) — lower = student score distribution resembles teacher more.

Unlike mean comparisons, KL divergence captures **variance and shape** differences. A student can have the same mean as the teacher but a flatter or more skewed distribution, which KL will detect.

A positive KL reduction (Pre-FT → Post-FT) for a dimension means fine-tuning brought the score distribution closer to the teacher's on that dimension.

Outputs:
- Per-dimension table: KL(T∥Pre), KL(T∥Post), Reduction
- **V11** `kl_score_distributions.png` — 6-panel histogram grid (one per dimension)
- **V11b** `kl_summary_bar.png` — summary grouped bar (Pre vs Post KL per dimension)


````python
# ─── 12.16  P6 — KL Divergence on Score Distributions (All 6 Dimensions) ────
# Compares how score distributions differ between teacher and student stages.
# KL(P_teacher || Q_student): low = student distribution resembles teacher.
# Extends beyond mean comparisons to capture variance and shape differences.

import numpy as np
from scipy.stats import entropy
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def score_distribution(records: list, dim_key: str) -> np.ndarray:
    """Return smoothed probability vector over scores 1–5."""
    counts = np.zeros(5, dtype=float)

    for r in records:
        if r.get("status") == "ok":
            dim = (r.get("scores") or {}).get(dim_key, {})
            s   = dim.get("score") if isinstance(dim, dict) else dim

            if s is not None and 1 <= int(s) <= 5:
                counts[int(s) - 1] += 1.0

    counts += 1e-9          # Laplace smoothing
    return counts / counts.sum()

def kl_divergence(P: np.ndarray, Q: np.ndarray) -> float:
    """KL(P || Q) — how much Q deviates from P."""
    return float(entropy(P, Q))

def pretty_distribution(dist: np.ndarray) -> str:
    """
    Convert probability vector into readable percentage format:
    5 → 30%
    4 → 40%
    ...
    """
    lines = []

    for score, prob in zip(range(1, 6), dist):
        pct = prob * 100

        # skip extremely tiny smoothing values
        if pct >= 0.05:
            lines.append(f"{score} → {pct:.1f}%")

    return "\n".join(lines)

print("\n" + "="*72)
print("KL DIVERGENCE — Score Distributions per Dimension (All 6 Ds)")
print("="*72)

print(f"  KL(Teacher || Student) — lower = student distribution resembles teacher")
print(f"  KL = 0: identical distributions  |  >0.5: substantially different")

print()
print(f"  {'Dimension':<36} {'KL(T||Pre)':>12}  {'KL(T||Post)':>13}  {'KL Reduction':>13}")
print("  " + "─"*76)

_kl_results = {}

for key in DIM_KEYS:

    P  = score_distribution(teacher_scores,      key)
    Qb = score_distribution(student_pre_scores,  key)
    Qa = score_distribution(student_post_scores, key)

    kl_pre  = kl_divergence(P, Qb)
    kl_post = kl_divergence(P, Qa)
    kl_red  = kl_pre - kl_post

    _kl_results[key] = {
        "kl_pre": round(kl_pre, 6),
        "kl_post": round(kl_post, 6),
        "kl_reduction": round(kl_red, 6),

        "teacher_distribution": {
            str(i+1): round(float(P[i]), 4)
            for i in range(5)
        },

        "pre_ft_distribution": {
            str(i+1): round(float(Qb[i]), 4)
            for i in range(5)
        },

        "post_ft_distribution": {
            str(i+1): round(float(Qa[i]), 4)
            for i in range(5)
        },
    }

    interp = "✓ improved" if kl_red > 0 else "~ unchanged"

    print(f"  {DIM_LABELS[key]:<36} "
          f"{kl_pre:>12.6f}  "
          f"{kl_post:>13.6f}  "
          f"{kl_red:>+13.6f}  {interp}")

print()
print("  KL Reduction > 0 → fine-tuning moved student distribution closer to teacher.")
print("  Largest KL post-FT → dimension where distributions diverge most from teacher.")

# ─── Exact Score Distribution Printout ───────────────────────────────────────

print("\n" + "="*110)
print("V11_kl_score_distributions — Exact Score Distributions")
print("="*110)

for key in DIM_KEYS:

    P  = score_distribution(teacher_scores,      key)
    Qb = score_distribution(student_pre_scores,  key)
    Qa = score_distribution(student_post_scores, key)

    print(f"\n{DIM_LABELS[key]}")
    print("─" * 110)

    print(f"{'Score':<8} {'Teacher':>14} {'Pre-FT':>14} {'Post-FT':>14}")
    print(" " + "─"*55)

    # Print from 5 → 1
    for score in range(5, 0, -1):

        idx = score - 1

        teacher_pct = P[idx]  * 100
        pre_pct     = Qb[idx] * 100
        post_pct    = Qa[idx] * 100

        print(
            f"{score:<8} "
            f"{teacher_pct:>11.1f}%   "
            f"{pre_pct:>11.1f}%   "
            f"{post_pct:>11.1f}%"
        )

# ── Heatmap of score distributions (6 dims × 5 score levels) ─────────────────

fig = plt.figure(figsize=(16, 10))
gs  = gridspec.GridSpec(2, 3, figure=fig, hspace=0.45, wspace=0.35)

score_labels = ["1", "2", "3", "4", "5"]
x = np.arange(5)
width = 0.28

for idx, key in enumerate(DIM_KEYS):

    row, col = divmod(idx, 3)
    ax = fig.add_subplot(gs[row, col])

    P  = score_distribution(teacher_scores,      key)
    Qb = score_distribution(student_pre_scores,  key)
    Qa = score_distribution(student_post_scores, key)

    ax.bar(x - width, P,  width,
           color="#2563EB", label="Teacher", alpha=0.8)

    ax.bar(x, Qb, width,
           color="#DC2626", label="Pre-FT", alpha=0.8)

    ax.bar(x + width, Qa, width,
           color="#16A34A", label="Post-FT", alpha=0.8)

    kl_v = _kl_results[key]["kl_post"]

    ax.set_title(
        f"{DIM_LABELS[key]}\nKL(T||Post)={kl_v:.4f}",
        fontsize=9
    )

    ax.set_xticks(x)
    ax.set_xticklabels(score_labels)

    ax.set_xlabel("Score (1–5)", fontsize=8)
    ax.set_ylabel("Proportion", fontsize=8)

    ax.grid(axis="y", alpha=0.3)

    if idx == 0:
        ax.legend(fontsize=7)

plt.suptitle(
    "Score Distribution per Dimension — KL Divergence Analysis\n"
    "(Teacher vs Pre-FT vs Post-FT Student)",
    fontsize=12,
    y=1.01
)

_p7_path = os.path.join(PLOT_DIR, "V11_kl_score_distributions.png")

plt.savefig(_p7_path, dpi=150, bbox_inches="tight")
plt.close()

print(f"\n✓ KL divergence distribution chart saved → {_p7_path}")

# ── Summary KL bar chart ──────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(11, 4))

dim_short = [
    k.split("_", 1)[1].replace("_", "\n").title()
    for k in DIM_KEYS
]

kl_pre_vals  = [_kl_results[k]["kl_pre"]  for k in DIM_KEYS]
kl_post_vals = [_kl_results[k]["kl_post"] for k in DIM_KEYS]

x2 = np.arange(len(DIM_KEYS))

ax.bar(
    x2 - 0.2,
    kl_pre_vals,
    0.38,
    color="#DC2626",
    label="KL(Teacher || Pre-FT)",
    alpha=0.85
)

ax.bar(
    x2 + 0.2,
    kl_post_vals,
    0.38,
    color="#16A34A",
    label="KL(Teacher || Post-FT)",
    alpha=0.85
)

ax.set_xticks(x2)
ax.set_xticklabels(dim_short, fontsize=9)

ax.set_ylabel("KL Divergence", fontsize=11)

ax.set_title(
    "KL Divergence per Dimension (Teacher || Student)\n"
    "Lower Post-FT bars = fine-tuning brought score distributions closer to teacher",
    fontsize=11
)

ax.legend(fontsize=10)
ax.grid(axis="y", alpha=0.3)

plt.tight_layout()

_p7b_path = os.path.join(PLOT_DIR, "V11b_kl_summary_bar.png")

plt.savefig(_p7b_path, dpi=150)
plt.close()

print(f"✓ KL summary bar chart saved → {_p7b_path}")

# ── Save ──────────────────────────────────────────────────────────────────────

_stats_extended["kl_divergence"] = _kl_results

with open(COMPARISON_PATH, "w", encoding="utf-8") as f:
    json.dump(_stats_extended, f, indent=2, ensure_ascii=False)

print(f"✓ KL divergence results saved → {COMPARISON_PATH}")
````

### 12.17  Perplexity — Student Model Fluency on MCQ Text Sources

Computes the **token-level perplexity** of the post-FT student model on three text sources:

| Source | What low perplexity means |
|---|---|
| Teacher MCQs | Student finds teacher-generated text natural (style alignment) |
| Student Pre-FT MCQs | Baseline — how natural the base model's own outputs are |
| Student Post-FT MCQs | How natural the fine-tuned model's own outputs are |

A lower perplexity on teacher MCQs after fine-tuning (compared to before) indicates that the student has **internalised the teacher's generation style**, not just its scoring rubric.

The **style gap %** quantifies the residual difference: `(PPL_on_teacher − PPL_on_own) / PPL_on_own × 100`.

> This metric is **model-based and judge-free** — it requires the post-FT model to be loaded (or loadable from `KAGGLE_BEST_MERGED_DIR`).

Outputs:
- Perplexity table (mean, std, n per source)
- Style gap interpretation
- **V12** `perplexity.png` — bar chart of mean PPL per text source


````python
# ─── 12.17  Perplexity — Student Model Fluency ────────────────────────────────
import torch, math, gc, os, json
import numpy as np
from transformers import AutoModelForCausalLM, AutoTokenizer

POST_FT_HF_REPO = "Rehab-Hamdy/mcq-bloom-qwen-merged_v4"

# ── Load score records from disk if not already in scope / empty ──────────────
if "teacher_scores" not in dir() or not teacher_scores:
    teacher_scores = load_jsonl_safe(TEACHER_SCORES_PATH)
    print(f"Loaded teacher_scores from disk: {len(teacher_scores):,} records")

if "student_pre_scores" not in dir() or not student_pre_scores:
    student_pre_scores = load_jsonl_safe(STUDENT_PRE_SCORES_PATH)
    print(f"Loaded student_pre_scores from disk: {len(student_pre_scores):,} records")

if "student_post_scores" not in dir() or not student_post_scores:
    student_post_scores = load_jsonl_safe(STUDENT_POST_SCORES_PATH)
    print(f"Loaded student_post_scores from disk: {len(student_post_scores):,} records")

print(f"Score counts → teacher: {len(teacher_scores)}, pre-FT: {len(student_pre_scores)}, post-FT: {len(student_post_scores)}")

# ── FIX: Build chunk_key → mcq_text lookup from the OUTPUT files ──────────────
# Judge score records only contain scores, not MCQ text. The actual question +
# options live in the original output JSONL files (TEACHER_OUTPUT_PATH etc.).
def _build_text_lookup(output_path: str) -> dict:
    """Build {chunk_id_slotindex: 'question optA optB ...'} from an output JSONL."""
    lookup = {}
    for r in load_jsonl_safe(output_path):
        mcq  = r.get("mcq") or {}
        q    = mcq.get("question", "").strip()
        opts = mcq.get("options") or {}
        if not q:
            continue
        opts_text = " ".join(f"{k}: {v}" for k, v in opts.items())
        key = f"{r.get('chunk_id', '')}_{r.get('slot_index', '')}"
        lookup[key] = f"{q} {opts_text}".strip()
    return lookup

print("\nBuilding MCQ text lookups from output files...")
_teacher_text_lookup   = _build_text_lookup(TEACHER_OUTPUT_PATH)
_pre_ft_text_lookup    = _build_text_lookup(STUDENT_PRE_OUTPUT_PATH)
_post_ft_text_lookup   = _build_text_lookup(STUDENT_POST_OUTPUT_PATH)
print(f"  teacher_outputs   : {len(_teacher_text_lookup):,} entries  [{TEACHER_OUTPUT_PATH}]")
print(f"  pre_ft_outputs    : {len(_pre_ft_text_lookup):,} entries  [{STUDENT_PRE_OUTPUT_PATH}]")
print(f"  post_ft_outputs   : {len(_post_ft_text_lookup):,} entries  [{STUDENT_POST_OUTPUT_PATH}]")


def _texts_from_score_records(score_records: list, text_lookup: dict, max_n: int = 100) -> list:
    """
    Extract MCQ texts for perplexity computation.
    Uses the score records (for status=ok filtering) and a pre-built
    text lookup keyed by chunk_id_slotindex (from the output JSONL files).
    """
    texts = []
    for r in score_records:
        if r.get("status") != "ok":
            continue
        key  = f"{r.get('chunk_id', '')}_{r.get('slot_index', '')}"
        text = text_lookup.get(key, "").strip()
        if text:
            texts.append(text)
        if len(texts) >= max_n:
            break
    return texts


def compute_perplexity(model, tokenizer, texts: list,
                       max_length: int = 256, device: str = "cuda") -> dict:
    model.eval()
    nlls = []
    with torch.no_grad():
        for text in texts:
            enc = tokenizer(text, return_tensors="pt",
                            max_length=max_length, truncation=True)
            input_ids = enc["input_ids"].to(device)
            if input_ids.shape[1] < 2:
                continue
            labels = input_ids.clone()
            out    = model(input_ids=input_ids, labels=labels)
            nll    = out.loss.item()
            if math.isfinite(nll):
                nlls.append(nll)

    if not nlls:
        return {"mean_ppl": None, "std_ppl": None, "n": 0}
    nlls_arr = np.array(nlls)
    ppls     = np.exp(nlls_arr)
    return {
        "mean_ppl":  round(float(np.mean(ppls)),     4),
        "std_ppl":   round(float(np.std(ppls)),      4),
        "mean_nll":  round(float(np.mean(nlls_arr)), 6),
        "n":         len(nlls),
    }


# ── Load post-FT model ────────────────────────────────────────────────────────
_ppl_device       = "cuda" if torch.cuda.is_available() else "cpu"
_ppl_model        = None
_ppl_tokenizer    = None
_ppl_model_source = None
_ppl_loaded_here  = False

for _mvar, _tvar, _src in [
    ("post_model", "post_tokenizer", "already_loaded_post"),
    ("pre_model",  "pre_tokenizer",  "already_loaded_pre"),
]:
    if _mvar in dir() and _tvar in dir():
        _ppl_model        = eval(_mvar)
        _ppl_tokenizer    = eval(_tvar)
        _ppl_model_source = _src
        print(f"Using already-loaded model: {_mvar}")
        break

if _ppl_model is None:
    for _path in [KAGGLE_BEST_MERGED_DIR, KAGGLE_MERGED_DIR]:
        if os.path.exists(_path):
            print(f"Loading post-FT model from local path: {_path}")
            try:
                _ppl_tokenizer = AutoTokenizer.from_pretrained(
                    _path, token=HF_TOKEN, trust_remote_code=True)
                _ppl_model = AutoModelForCausalLM.from_pretrained(
                    _path, torch_dtype=torch.bfloat16,
                    device_map="auto", token=HF_TOKEN, trust_remote_code=True)
                _ppl_model.eval()
                _ppl_model_source = _path
                _ppl_loaded_here  = True
                print(f"  Model loaded from local path.")
                break
            except Exception as _e:
                print(f"  Failed to load from {_path}: {_e}")

if _ppl_model is None:
    print(f"Loading post-FT model from HF: {POST_FT_HF_REPO}")
    try:
        _ppl_tokenizer = AutoTokenizer.from_pretrained(
            POST_FT_HF_REPO, token=HF_TOKEN, trust_remote_code=True)
        _ppl_model = AutoModelForCausalLM.from_pretrained(
            POST_FT_HF_REPO, torch_dtype=torch.bfloat16,
            device_map="auto", token=HF_TOKEN, trust_remote_code=True)
        _ppl_model.eval()
        _ppl_model_source = POST_FT_HF_REPO
        _ppl_loaded_here  = True
        print(f"  Model loaded from HF Hub: {POST_FT_HF_REPO}")
    except Exception as _e:
        print(f"  Failed to load from HF: {_e}")

if _ppl_model is None:
    print("WARNING: Post-FT model could not be loaded. Skipping perplexity computation.")
else:
    PPL_MAX_N = 150

    teacher_texts  = _texts_from_score_records(teacher_scores,      _teacher_text_lookup,  PPL_MAX_N)
    pre_ft_texts   = _texts_from_score_records(student_pre_scores,  _pre_ft_text_lookup,   PPL_MAX_N)
    post_ft_texts  = _texts_from_score_records(student_post_scores, _post_ft_text_lookup,  PPL_MAX_N)

    print(f"\nTexts extracted → teacher: {len(teacher_texts)}, pre-FT: {len(pre_ft_texts)}, post-FT: {len(post_ft_texts)}")

    print(f"\nComputing perplexity on {len(teacher_texts)} teacher MCQs...")
    ppl_on_teacher = compute_perplexity(_ppl_model, _ppl_tokenizer, teacher_texts, device=_ppl_device)

    print(f"Computing perplexity on {len(pre_ft_texts)} pre-FT MCQs...")
    ppl_on_pre_ft  = compute_perplexity(_ppl_model, _ppl_tokenizer, pre_ft_texts,  device=_ppl_device)

    print(f"Computing perplexity on {len(post_ft_texts)} post-FT MCQs...")
    ppl_on_post_ft = compute_perplexity(_ppl_model, _ppl_tokenizer, post_ft_texts, device=_ppl_device)

    print("\n" + "=" * 65)
    print("PERPLEXITY — Post-FT Student Model on MCQ Texts")
    print("=" * 65)
    print(f"  Model   : {_ppl_model_source}")
    print(f"  Metric  : mean token-level perplexity (exp of mean NLL per token)")
    print(f"  Lower   : model finds the text more 'natural' / fluent")
    print()
    print(f"  {'Text Source':<30} {'Mean PPL':>10}  {'Std PPL':>9}  {'Mean NLL':>10}  {'n':>5}")
    print("  " + "─" * 65)

    for label, res in [
        ("Teacher MCQs",         ppl_on_teacher),
        ("Student Pre-FT MCQs",  ppl_on_pre_ft),
        ("Student Post-FT MCQs", ppl_on_post_ft),
    ]:
        if res["mean_ppl"] is not None:
            print(f"  {label:<30} {res['mean_ppl']:>10.2f}  {res['std_ppl']:>9.2f}  "
                  f"{res['mean_nll']:>10.4f}  {res['n']:>5}")
        else:
            print(f"  {label:<30} {'N/A':>10}")

    print()
    if ppl_on_teacher["mean_ppl"] and ppl_on_pre_ft["mean_ppl"]:
        delta_pre  = ppl_on_pre_ft["mean_ppl"]  - ppl_on_teacher["mean_ppl"]
        delta_post = (ppl_on_post_ft["mean_ppl"] - ppl_on_teacher["mean_ppl"]
                      if ppl_on_post_ft["mean_ppl"] else None)
        print(f"  PPL difference (Pre-FT  − Teacher) : {delta_pre:+.2f}")
        if delta_post is not None:
            print(f"  PPL difference (Post-FT − Teacher) : {delta_post:+.2f}")
            if delta_post < delta_pre:
                print("  Post-FT student outputs are closer to teacher style than pre-FT (expected).")
            if ppl_on_post_ft["mean_ppl"]:
                gap_pct = (ppl_on_teacher["mean_ppl"] / ppl_on_post_ft["mean_ppl"] - 1) * 100
                print(f"  Student finds teacher MCQs {abs(gap_pct):.1f}% "
                      f"{'more' if gap_pct < 0 else 'less'} surprising than its own outputs.")

    import matplotlib.pyplot as plt
    _ppl_labels = ["Teacher\nMCQs", "Pre-FT\nMCQs", "Post-FT\nMCQs"]
    _ppl_vals   = [ppl_on_teacher["mean_ppl"] or 0,
                   ppl_on_pre_ft["mean_ppl"]  or 0,
                   ppl_on_post_ft["mean_ppl"] or 0]
    _ppl_stds   = [ppl_on_teacher["std_ppl"]  or 0,
                   ppl_on_pre_ft["std_ppl"]   or 0,
                   ppl_on_post_ft["std_ppl"]  or 0]
    _ppl_colors = ["#2563EB", "#DC2626", "#16A34A"]

    fig, ax = plt.subplots(figsize=(7, 4))
    bars = ax.bar(_ppl_labels, _ppl_vals, yerr=_ppl_stds, capsize=5,
                  color=_ppl_colors, alpha=0.85, width=0.45,
                  error_kw={"elinewidth": 2})
    for bar, v in zip(bars, _ppl_vals):
        ax.text(bar.get_x() + bar.get_width() / 2, v + max(_ppl_vals) * 0.02,
                f"{v:.1f}", ha="center", va="bottom", fontsize=11, fontweight="bold")
    ax.set_ylabel("Mean Perplexity (lower = more fluent)", fontsize=11)
    ax.set_title(
        f"Post-FT Student Perplexity on Different MCQ Text Sources\n"
        f"Model: {_ppl_model_source}",
        fontsize=10,
    )
    ax.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    _ppl_path = os.path.join(PLOT_DIR, "V12_perplexity.png")
    plt.savefig(_ppl_path, dpi=150)
    plt.close()
    print(f"\n  Perplexity chart saved -> {_ppl_path}")

    _stats_extended["perplexity"] = {
        "model_source":        str(_ppl_model_source),
        "ppl_on_teacher_mcqs": ppl_on_teacher,
        "ppl_on_pre_ft_mcqs":  ppl_on_pre_ft,
        "ppl_on_post_ft_mcqs": ppl_on_post_ft,
    }
    with open(COMPARISON_PATH, "w", encoding="utf-8") as f:
        json.dump(_stats_extended, f, indent=2, ensure_ascii=False)
    print(f"  Perplexity results saved -> {COMPARISON_PATH}")

    if _ppl_loaded_here:
        del _ppl_model, _ppl_tokenizer
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        print("  Perplexity model unloaded from GPU.")
````

````python
# ─── 12.19  Full Evaluation Summary & Final Comparison ──────────────────────
# Aggregates every evaluation signal from sections 12.9 – 12.18 into a single
# structured report and saves the final master JSON.
#
# Signals consolidated:
#   • CMQS (mean ± std) for Teacher / Pre-FT / Post-FT
#   • Per-dimension scores (D1–D6) with FT Gain and QPS
#   • Quality Preservation Score (QPS) and FT Gain
#   • Bloom-level CMQS breakdown
#   • Statistical significance (Wilcoxon + Cohen's d)
#   • Sentence-BERT semantic similarity
#   • Dimension gap ranking (failure mode analysis)
#   • Retention Score & efficiency analysis
#   • KL divergence per dimension
#   • Perplexity (student fluency on teacher vs. student MCQs)

import json, os, math
import numpy as np

SEPARATOR = "═" * 72

def _safe(d, *keys, default="-"):
    """Safely navigate nested dicts."""
    v = d
    for k in keys:
        if not isinstance(v, dict):
            return default
        v = v.get(k, default)
    return v

def _pct(val, default="-"):
    if isinstance(val, (int, float)) and not math.isnan(val):
        return f"{val:.1f}%"
    return default

def _f(val, fmt=".3f", default="-"):
    if isinstance(val, (int, float)) and not math.isnan(val):
        return format(val, fmt)
    return default

# ── Load the master comparison JSON (built incrementally by 12.9 – 12.18) ────
if os.path.exists(COMPARISON_PATH):
    with open(COMPARISON_PATH, encoding="utf-8") as _f_:
        _all = json.load(_f_)
else:
    # Fallback: build from in-memory stats if JSON not yet written
    _all = {
        "teacher":         stats_teacher,
        "student_pre_ft":  stats_pre_ft,
        "student_post_ft": stats_post_ft,
        "summary": {
            "QPS_overall_pct":   overall_qps,
            "FT_gain_CMQS":      ft_gain,
            "teacher_CMQS_mean": teacher_cmqs,
            "pre_ft_CMQS_mean":  pre_ft_cmqs,
            "post_ft_CMQS_mean": post_ft_cmqs,
        }
    }

_t    = _all.get("teacher",            {})
_pre  = _all.get("student_pre_ft",     {})
_pst  = _all.get("student_post_ft",    {})
_smry = _all.get("summary",            {})
_sig  = _all.get("significance_tests", {})
_sbert= _all.get("semantic_similarity",{})
_gaps = _all.get("gap_ranking",        [])
_rs   = _all.get("retention_score",    {})
_kl   = _all.get("kl_divergence",      {})
_ppl  = _all.get("perplexity",         {})

print()
print("╔" + SEPARATOR + "╗")
print("║" + "  SECTION 12 — FULL EVALUATION SUMMARY & FINAL COMPARISON".center(72) + "║")
print("╚" + SEPARATOR + "╝")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 1 — Sample Counts
# ─────────────────────────────────────────────────────────────────────────────
print()
print("┌─ 1. Sample Coverage " + "─"*51 + "┐")
print(f"  Teacher MCQs scored   : {_t.get('n_valid', '-'):>6}  (of {_t.get('n_total', '-')} submitted)")
print(f"  Pre-FT  MCQs scored   : {_pre.get('n_valid', '-'):>6}  (of {_pre.get('n_total', '-')} submitted)")
print(f"  Post-FT MCQs scored   : {_pst.get('n_valid', '-'):>6}  (of {_pst.get('n_total', '-')} submitted)")
print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 2 — CMQS Core Results
# ─────────────────────────────────────────────────────────────────────────────
print()
print("┌─ 2. Core CMQS Results " + "─"*49 + "┐")
t_cmqs  = _safe(_t,   "CMQS", "mean", default=0.0)
b_cmqs  = _safe(_pre, "CMQS", "mean", default=0.0)
a_cmqs  = _safe(_pst, "CMQS", "mean", default=0.0)
t_std   = _safe(_t,   "CMQS", "std",  default=0.0)
b_std   = _safe(_pre, "CMQS", "std",  default=0.0)
a_std   = _safe(_pst, "CMQS", "std",  default=0.0)
_qps_pre  = round(b_cmqs / t_cmqs * 100, 1) if t_cmqs else None
_qps_post = round(a_cmqs / t_cmqs * 100, 1) if t_cmqs else None
_ft_gain  = round(a_cmqs - b_cmqs, 4) if b_cmqs else None

print(f"  {'Stage':<28} {'CMQS Mean':>10}  {'Std':>6}  {'QPS':>8}  {'vs Prev':>9}")
print("  " + "─"*56)
print(f"  {'Teacher (Gemini)':<28} {_f(t_cmqs):>10}  {_f(t_std):>6}  {'100.0%':>8}  {'—':>9}")
print(f"  {'Student Pre-FT (Base Qwen)':<28} {_f(b_cmqs):>10}  {_f(b_std):>6}  {_pct(_qps_pre):>8}  {'—':>9}")
print(f"  {'Student Post-FT (Distilled)':<28} {_f(a_cmqs):>10}  {_f(a_std):>6}  {_pct(_qps_post):>8}  {f'{_ft_gain:+.4f}' if _ft_gain else '—':>9}  ←★")
print()

_qps_overall = _smry.get("QPS_overall_pct", _qps_post)
if isinstance(_qps_overall, (int, float)):
    if   _qps_overall >= 90: _qps_verdict = "EXCELLENT — near teacher quality"
    elif _qps_overall >= 80: _qps_verdict = "GOOD — effective distillation"
    elif _qps_overall >= 70: _qps_verdict = "MODERATE — acceptable"
    elif _qps_overall >= 60: _qps_verdict = "WEAK — review pipeline"
    else:                    _qps_verdict = "POOR — distillation failed"
    print(f"  Quality Preservation Score (QPS) : {_pct(_qps_overall)}  →  {_qps_verdict}")
print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 3 — Per-Dimension Comparison
# ─────────────────────────────────────────────────────────────────────────────
print()
print("┌─ 3. Per-Dimension Scores (D1–D6 + CMQS) " + "─"*30 + "┐")
print(f"  {'Dimension':<36} {'Teacher':>8} {'Pre-FT':>8} {'Post-FT':>9} {'FT Gain':>9} {'Post-QPS':>9}")
print("  " + "─"*80)

_ALL_KEYS   = DIM_KEYS + ["CMQS"]
_ALL_LABELS = {**DIM_LABELS, "CMQS": "CMQS (Composite, weighted)"}

for key in _ALL_KEYS:
    tm   = _safe(_t,   key, "mean", default=0.0)
    bm   = _safe(_pre, key, "mean", default=0.0)
    am   = _safe(_pst, key, "mean", default=0.0)
    gain = round(am - bm, 3) if isinstance(am, float) and isinstance(bm, float) else None
    pqps = round(am / tm * 100, 1) if isinstance(tm, float) and tm else None
    star = "  ←★" if key == "CMQS" else ""
    print(f"  {_ALL_LABELS[key]:<36} {_f(tm):>8} {_f(bm):>8} {_f(am):>9} "
          f"{f'{gain:+.3f}' if gain is not None else '—':>9} {_pct(pqps):>9}{star}")
print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 4 — Bloom-Level CMQS Breakdown
# ─────────────────────────────────────────────────────────────────────────────
print()
print("┌─ 4. Bloom-Level CMQS Breakdown " + "─"*40 + "┐")
print(f"  {'Level':<13} {'Teacher':>9} {'Pre-FT':>9} {'Post-FT':>10} {'FT Gain':>9} {'Post-QPS':>9}")
print("  " + "─"*62)
_BLOOM_LEVELS = ["remember", "understand", "apply", "analyze", "evaluate", "create"]
for bl in _BLOOM_LEVELS:
    tm   = _safe(_t,   "bloom_breakdown", bl, "mean", default=None)
    bm   = _safe(_pre, "bloom_breakdown", bl, "mean", default=None)
    am   = _safe(_pst, "bloom_breakdown", bl, "mean", default=None)
    gain = round(am - bm, 3) if isinstance(am, float) and isinstance(bm, float) else None
    pqps = round(am / tm * 100, 1) if isinstance(tm, float) and tm and isinstance(am, float) else None
    print(f"  {bl:<13} {_f(tm):>9} {_f(bm):>9} {_f(am):>10} "
          f"{f'{gain:+.3f}' if gain is not None else '—':>9} {_pct(pqps):>9}")
print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 5 — Statistical Significance (Wilcoxon + Cohen's d)
# ─────────────────────────────────────────────────────────────────────────────
if _sig:
    print()
    print("┌─ 5. Statistical Significance — Wilcoxon Signed-Rank + Cohen's d " + "─"*5 + "┐")
    print(f"  α = 0.05  |  H0: no difference  |  one-sided (Post-FT > Pre-FT)")
    print()
    print(f"  {'Comparison / Dimension':<36} {'p-value':>10}  {'d':>7}  {'Effect':>10}  {'Sig?':>10}")
    print("  " + "─"*72)

    def _effect(d):
        ad = abs(d) if isinstance(d, float) else 0
        if ad < 0.2: return "negligible"
        if ad < 0.5: return "small"
        if ad < 0.8: return "medium"
        return "large"

    for comp_label, comp_key in [
        ("CMQS  Post-FT > Pre-FT",  "cmqs_post_vs_pre"),
        ("CMQS  Teacher > Post-FT", "cmqs_teacher_vs_post"),
    ]:
        row = _sig.get(comp_key, {})
        p   = row.get("p")
        d   = row.get("cohens_d", 0.0)
        n   = row.get("n", "-")
        sig = "✓ sig" if isinstance(p, float) and p < 0.05 else "✗ n.s."
        print(f"  {comp_label:<36} {_f(p, '.4f'):>10}  {_f(d, '+.3f'):>7}  {_effect(d):>10}  {sig:>10}  n={n}")

    print()
    for key in DIM_KEYS:
        row = _sig.get(key, {})
        p   = row.get("p")
        d   = row.get("cohens_d", 0.0)
        n   = row.get("n", "-")
        sig = "✓ sig" if isinstance(p, float) and p < 0.05 else "✗ n.s."
        print(f"  {DIM_LABELS[key]:<36} {_f(p, '.4f'):>10}  {_f(d, '+.3f'):>7}  {_effect(d):>10}  {sig:>10}  n={n}")
    print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 6 — Semantic Similarity (Sentence-BERT)
# ─────────────────────────────────────────────────────────────────────────────
if _sbert:
    print()
    print("┌─ 6. Sentence-BERT Semantic Similarity (Teacher ↔ Student) " + "─"*12 + "┐")
    pre_s  = _sbert.get("pre_ft",  {})
    post_s = _sbert.get("post_ft", {})
    print(f"  {'Pair':<30} {'Mean Sim':>10}  {'Std':>7}  {'n':>6}")
    print("  " + "─"*52)
    print(f"  {'Teacher ↔ Pre-FT Student':<30} {_f(pre_s.get('mean'),  '.4f'):>10}  {_f(pre_s.get('std'),  '.4f'):>7}  {pre_s.get('n', '-'):>6}")
    print(f"  {'Teacher ↔ Post-FT Student':<30} {_f(post_s.get('mean'), '.4f'):>10}  {_f(post_s.get('std'), '.4f'):>7}  {post_s.get('n', '-'):>6}")
    delta = None
    if isinstance(post_s.get("mean"), float) and isinstance(pre_s.get("mean"), float):
        delta = round(post_s["mean"] - pre_s["mean"], 4)
    print(f"  Similarity gain (Post − Pre)  : {_f(delta, '+.4f') if delta is not None else '—'}")
    interp = "Fine-tuning improved alignment with teacher." if (delta and delta > 0) else "No gain in alignment."
    print(f"  Interpretation                : {interp}")
    print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 7 — Dimension Gap Ranking (Failure Mode Analysis)
# ─────────────────────────────────────────────────────────────────────────────
if _gaps:
    print()
    print("┌─ 7. Dimension Gap Ranking — Knowledge Transfer Failure Modes " + "─"*10 + "┐")
    print(f"  Gap = Teacher − Post-FT  |  Gain = Post-FT − Pre-FT  (ranked largest→smallest gap)")
    print()
    print(f"  {'Dimension':<36} {'Gap (↓)':>9} {'Gain (↑)':>9} {'Post-QPS':>9}  Severity")
    print("  " + "─"*72)
    for row in _gaps:
        gap  = row.get("gap",      0)
        gain = row.get("gain",     0)
        pqps = row.get("post_qps", 0)
        lbl  = row.get("label",    row.get("key", ""))
        if   gap < 0.25: sev = "✓ negligible"
        elif gap < 0.50: sev = "~ small"
        elif gap < 0.75: sev = "! moderate"
        else:            sev = "✗ LARGE"
        print(f"  {lbl:<36} {gap:>+9.3f} {gain:>+9.3f} {_pct(pqps):>9}  {sev}")
    print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 8 — KL Divergence per Dimension
# ─────────────────────────────────────────────────────────────────────────────
if _kl:
    print()
    print("┌─ 8. KL Divergence per Dimension (Teacher || Student, score distributions) ┐")
    print(f"  Lower KL post-FT = fine-tuning moved score distribution closer to teacher")
    print()
    print(f"  {'Dimension':<36} {'KL(T||Pre)':>12}  {'KL(T||Post)':>12}  {'Reduction':>12}")
    print("  " + "─"*72)
    for key in DIM_KEYS:
        row     = _kl.get(key, {})
        kl_pre  = row.get("kl_pre",       None)
        kl_post = row.get("kl_post",      None)
        kl_red  = row.get("kl_reduction", None)
        interp  = " ✓ improved" if isinstance(kl_red, float) and kl_red > 0 else "~ unchanged"
        print(f"  {DIM_LABELS[key]:<36} {_f(kl_pre, '.6f'):>12}  {_f(kl_post, '.6f'):>12}  "
              f"{_f(kl_red, '+.6f') if isinstance(kl_red, float) else '—':>12}  {interp}")
    print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 9 — Retention Score & Efficiency
# Keys written by cell 12.15:
#   teacher_latency_ms, student_latency_ms, latency_ratio_T_over_S,
#   speed_description, cmqs_retention_pct, retention_score_RS,
#   latency_source, latency_sample{}
# NOTE: compression_ratio / latency_speedup / *_params_B are NOT saved by
#       12.15 — read latency_ratio_T_over_S and speed_description instead.
# ─────────────────────────────────────────────────────────────────────────────
if _rs:
    print()
    print("┌─ 9. Retention Score & Efficiency Analysis " + "─"*29 + "┐")
    print(f"  RS = (Post-FT CMQS / Teacher CMQS) × (Teacher Latency / Student Latency)")
    print(f"  RS > 1.0 → student delivers better quality-per-compute than teacher")
    print()
    print(f"  {'Metric':<38} {'Teacher':>12}  {'Post-FT Student':>16}")
    print("  " + "─"*68)
    print(f"  {'Inference Latency (ms)':<38} {_f(_rs.get('teacher_latency_ms'), '.1f'):>12}  {_f(_rs.get('student_latency_ms'), '.1f'):>16}")
    print(f"  {'CMQS (mean)':<38} {_f(t_cmqs):>12}  {_f(a_cmqs):>16}")
    print()

    # latency_ratio_T_over_S: >1.0 → student faster, <1.0 → student slower
    _lat_ratio = _rs.get("latency_ratio_T_over_S")
    if isinstance(_lat_ratio, float):
        if _lat_ratio >= 1.0:
            _speedup_str = f"{_lat_ratio:.2f}× student faster"
        else:
            _speedup_str = f"{1 / _lat_ratio:.2f}× student slower"
    else:
        _speedup_str = "—"

    print(f"  Latency Ratio (T/S)      : {_f(_lat_ratio, '.4f')}  ({_speedup_str})")
    print(f"  Speed Summary            : {_rs.get('speed_description', '—')}")
    print(f"  CMQS Retention           : {_pct(_rs.get('cmqs_retention_pct'))}")
    print(f"  Retention Score (RS)     : {_f(_rs.get('retention_score_RS'), '.4f')}")

    rs_val = _rs.get("retention_score_RS", 0)
    if isinstance(rs_val, float):
        verdict = "BETTER quality-per-compute than teacher" if rs_val >= 1.0 else "below teacher quality-per-compute"
        print(f"  Verdict                  : RS = {rs_val:.2f} → Student achieves {verdict}")

    # Structured sample details from latency_sample sub-dict
    _ls = _rs.get("latency_sample", {})
    if _ls:
        print()
        _ls_desc = (
            f"chunk_id={_ls.get('chunk_id', '?')} | "
            f"bloom={_ls.get('bloom_level', '?')} | "
            f"chunk_type={_ls.get('chunk_type', '?')} | "
            f"concept='{_ls.get('concept', '?')}'"
        )
        print(f"  Latency sample           : {_ls_desc}")
        print(f"  Student model            : {_ls.get('student_model', '—')}")
        print(f"  Teacher model            : {_ls.get('teacher_model', '—')}")
        print(f"  Repetitions              : {_ls.get('n_warmup', '?')} warmup + {_ls.get('n_measure', '?')} timed runs")
    print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 10 — Perplexity
# ─────────────────────────────────────────────────────────────────────────────
if _ppl:
    print()
    print("┌─ 10. Perplexity — Post-FT Student Fluency on MCQ Text Sources " + "─"*8 + "┐")
    print(f"  Lower PPL = student finds the text more natural / in-distribution")
    print()
    print(f"  {'Text Source':<30} {'Mean PPL':>10}  {'Std PPL':>9}  {'n':>6}")
    print("  " + "─"*56)
    for _label, _key in [("Teacher MCQs",       "ppl_on_teacher_mcqs"),
                          ("Pre-FT MCQs",        "ppl_on_pre_ft_mcqs"),
                          ("Post-FT MCQs (own)", "ppl_on_post_ft_mcqs")]:
        r = _ppl.get(_key, {})
        print(f"  {_label:<30} {_f(r.get('mean_ppl'), '.2f'):>10}  {_f(r.get('std_ppl'), '.2f'):>9}  {r.get('n', '-'):>6}")
    print()
    _t_ppl = (_ppl.get("ppl_on_teacher_mcqs", {}) or {}).get("mean_ppl")
    _p_ppl = (_ppl.get("ppl_on_post_ft_mcqs", {}) or {}).get("mean_ppl")
    if isinstance(_t_ppl, float) and isinstance(_p_ppl, float):
        _style_gap = round((_p_ppl - _t_ppl) / _t_ppl * 100, 1) if _t_ppl else None
        print(f"  Style gap (PPL on Teacher − own): {_f(_style_gap, '+.1f')}%  — residual generation-style difference")
    print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 11 — Plots Reference
# ─────────────────────────────────────────────────────────────────────────────
print()
print("┌─ 11. Generated Evaluation Plots " + "─"*39 + "┐")
_plots = [
    ("V1",   "radar_all_stages.png",       "Spider chart: 6 dims × 3 stages"),
    ("V2",   "grouped_bar_dimensions.png", "Grouped bars per dimension (mean ± std)"),
    ("V3",   "bloom_level_lines.png",      "CMQS by Bloom level — line chart"),
    ("V4",   "quality_distribution.png",   "Quality label stacked bar"),
    ("V5",   "gap_heatmap.png",            "Gap heatmap: Teacher vs Student"),
    ("V6",   "cmqs_bar.png",              "Composite CMQS comparison bar"),
    ("V7",   "bloom_breakdown_bar.png",    "Bloom-level grouped bar with QPS labels"),
    ("V8",   "semantic_similarity.png",    "Sentence-BERT similarity distributions"),
    ("V9",   "gap_ranking.png",            "Horizontal bar: dimension gap ranking"),
    ("V10",  "retention_score.png",        "Efficiency–quality trade-off bars"),
    ("V11",  "kl_score_distributions.png", "Score distribution per dimension (KL)"),
    ("V11b",  "kl_summary_bar.png",         "KL divergence summary bar"),
    ("V12",  "perplexity.png",             "Student perplexity on MCQ sources"),
]
for tag, fname, desc in _plots:
    exists = "✓" if os.path.exists(os.path.join(PLOT_DIR, fname)) else "○"
    print(f"  {exists} {tag+'_'+fname:<38} {desc}")
print("└" + "─"*72 + "┘")

# ─────────────────────────────────────────────────────────────────────────────
# BLOCK 12 — Overall Verdict
# ─────────────────────────────────────────────────────────────────────────────
print()
print("╔" + SEPARATOR + "╗")
print("║" + "  OVERALL VERDICT".center(72) + "║")
print("╚" + SEPARATOR + "╝")
print()

_verdict_lines = []

# QPS verdict
_qps_v = _smry.get("QPS_overall_pct") or _qps_post
if isinstance(_qps_v, float):
    _verdict_lines.append(f"  [QPS]       Quality Preservation = {_pct(_qps_v)}  →  {_qps_verdict}")

# FT Gain verdict
_ftg = _smry.get("FT_gain_CMQS") or _ft_gain
if isinstance(_ftg, float):
    _dir = "improved" if _ftg > 0 else "degraded"
    _verdict_lines.append(f"  [FT Gain]   Fine-tuning {_dir} CMQS by {_ftg:+.4f} absolute")

# Semantic similarity verdict
if _sbert and isinstance(_sbert.get("post_ft", {}).get("mean"), float):
    _sim_post = _sbert["post_ft"]["mean"]
    _verdict_lines.append(f"  [SBERT]     Teacher↔Post-FT semantic similarity = {_sim_post:.4f}")


# Retention Score verdict — uses latency_ratio_T_over_S and speed_description
# (compression_ratio / latency_speedup are not saved by cell 12.15)
if _rs and isinstance(_rs.get("retention_score_RS"), float):
    _rs_v        = _rs["retention_score_RS"]
    _rs_ok       = "✓ > 1.0" if _rs_v >= 1.0 else "✗ < 1.0"
    _speed_desc  = _rs.get("speed_description", "—")
    _verdict_lines.append(
        f"  [RS]        Retention Score = {_rs_v:.4f}  ({_rs_ok}) — {_speed_desc}"
    )

# Statistical significance
_sig_cmqs = _sig.get("cmqs_post_vs_pre", {})
if _sig_cmqs:
    _p_sig   = _sig_cmqs.get("p")
    _d_sig   = _sig_cmqs.get("cohens_d", 0)
    _sig_str = f"p={_f(_p_sig, '.4f')}, d={_f(_d_sig, '+.3f')}"
    _sig_ok  = "✓ significant" if isinstance(_p_sig, float) and _p_sig < 0.05 else "✗ not significant"
    _verdict_lines.append(f"  [Sig]       CMQS Post>Pre Wilcoxon: {_sig_str}  →  {_sig_ok}")

for line in _verdict_lines:
    print(line)

print()
print(f"  {'Output file':}")
print(f"  {COMPARISON_PATH}")
print()
print("✓ Evaluation completed")
print()
````
